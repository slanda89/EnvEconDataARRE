# Download the 2024 SEEA Global Assessment Excel File

This stage downloads the official 2024 Global Assessment Excel file from the System of Environmental-Economic Accounting (SEEA) website.

The link provided opens the file through the Microsoft Office online viewer. Because this viewer link is not the direct Excel download link, the code first extracts the real file URL from the `src` parameter. After that, the code downloads the Excel file and saves it locally as:

`global_assessment_for_website_2024_final.xlsx`

After the file is downloaded, the code checks the available worksheet names so that the correct sheet can be selected in the next stage.

```python
import requests
from pathlib import Path
from urllib.parse import urlparse, parse_qs, unquote

# Office viewer link
viewer_url = "https://view.officeapps.live.com/op/view.aspx?src=https%3A%2F%2Fseea.un.org%2Fsites%2Fseea.un.org%2Ffiles%2Ffiles%2FGlobal_assessment%2F2024_GA%2Fglobal_assessment_for_website_2024_final.xlsx&wdOrigin=BROWSELINK"

# Extract the real Excel file URL from the src parameter
parsed_url = urlparse(viewer_url)
file_url = parse_qs(parsed_url.query)["src"][0]
file_url = unquote(file_url)

# Output file name
output_file = Path("global_assessment_for_website_2024_final.xlsx")

# Download the file
response = requests.get(file_url)
response.raise_for_status()

# Save file
output_file.write_bytes(response.content)

print(f"File downloaded successfully: {output_file}")
```
