#### a. UNSD Environmental Indicators Metadata and Country Availability Workbook


````python
# ============================================================
# UNSD Environmental Indicators Metadata + Country Availability
# ============================================================

import os
import re
import time
import requests
import pandas as pd
import numpy as np

from pathlib import Path
from bs4 import BeautifulSoup
from urllib.parse import urljoin, unquote, urlparse
from tqdm.auto import tqdm

from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

from openpyxl import load_workbook
from openpyxl.styles import Font, Alignment
from openpyxl.utils import get_column_letter
````

````python


# ============================================================
# 1. Paths and settings
# ============================================================

TEMPLATE_CSV = 'UN_unenv_final.csv'
BASE_PAGE_URL = 'https://unstats.un.org/unsd/envstats/qindicators'
DOWNLOAD_DIR = Path('unsd_env_indicator_files')
OUTPUT_XLSX = 'UNSD_Environmental_Indicators_Metadata_Availability.xlsx'

TIMEOUT = 60

DOWNLOAD_DIR.mkdir(exist_ok=True)

META_COLS = ['no', 'indicator', 'Definition', 'unit', 'source']
````

````python



# ============================================================
# 2. Load template file
# ============================================================

try:
    template_df = pd.read_csv(TEMPLATE_CSV, encoding='utf-8')
except UnicodeDecodeError:
    template_df = pd.read_csv(TEMPLATE_CSV, encoding='latin1')

country_cols = [col for col in template_df.columns if col not in META_COLS]

print(template_df.head())
print(template_df.shape)
print('Country columns:')
print(len(country_cols))
````

````python



# ============================================================
# 3. Basic helper functions
# ============================================================

def clean_text(value_obj):
    if value_obj is None:
        return ''
    text_val = str(value_obj)
    text_val = text_val.replace('\xa0', ' ')
    text_val = re.sub(r'\s+', ' ', text_val)
    return text_val.strip()


def safe_filename(name_val):
    name_val = unquote(str(name_val))
    name_val = re.sub(r'[<>:"/\\|?*]', '_', name_val)
    name_val = re.sub(r'\s+', ' ', name_val).strip()
    return name_val


def blank_like(value_obj):
    if pd.isna(value_obj):
        return True
    text_val = clean_text(value_obj).lower()
    return text_val in ['', '', 'none']


missing_tokens = {
    '',
    '...',
    '…',
    '..',
    'na',
    '',
    'n/a',
    'none',
    'not available',
    '-'
}


def is_reported_value(value_obj):
    if pd.isna(value_obj):
        return False

    text_val = clean_text(value_obj)

    if text_val.lower() in missing_tokens:
        return False

    return True


def normalize_header(value_obj):
    text_val = clean_text(value_obj)
    text_val = text_val.replace('\n', ' ')
    text_val = re.sub(r'\s+', ' ', text_val).strip()
    return text_val


def is_year_label(value_obj):
    text_val = clean_text(value_obj)

    if re.fullmatch(r'\d{4}', text_val):
        year_val = int(text_val)

        if 1800 <= year_val <= 2100:
            return True

    return False
````

````python
# ============================================================
# 4. Country-name matching helpers
# ============================================================

def norm_country_name(value_obj):
    text_val = clean_text(value_obj).lower()
    text_val = text_val.replace('&', 'and')
    text_val = text_val.replace('’', "'").replace('`', "'")
    text_val = re.sub(r'[\.,$$$$:;\-/]', ' ', text_val)
    text_val = re.sub(r'\s+', ' ', text_val).strip()
    return text_val


merge_groups = {
    'Bahamas': ['Bahamas', 'Bahamas, The'],
    'Bolivia': ['Bolivia', 'Bolivia (Plurinational State of)', 'Bolivia, Plurinational State of'],
    'British Virgin Islands': ['British Virgin Islands', 'Virgin Islands, British'],
    'Brunei Darussalam': ['Brunei Darussalam', 'Brunei'],
    'Cabo Verde': ['Cabo Verde', 'Cape Verde'],
    'Congo Dem. Rep.': ['Congo Dem. Rep.', 'Dem. Rep. of the Congo', 'Democratic Republic of the Congo', 'Congo, Dem. Rep.'],
    'Congo Rep.': ['Congo Rep.', 'Congo', 'Republic of Congo', 'Congo, Rep.'],
    'Côte dIvoire': ["Côte d'Ivoire", "Cote d'Ivoire", 'Côte dIvoire', 'Côte d’Ivoire'],
    'Curaçao': ['Curaçao', 'Curacao'],
    'Czech Republic': ['Czech Republic', 'Czechia'],
    'Eswatini': ['Eswatini', 'Kingdom of Eswatini', 'Swaziland'],
    'Hong Kong (S.A.R.)': ['Hong Kong (S.A.R.)', 'China, Hong Kong SAR', 'Hong Kong SAR', 'China, Hong Kong Special Administrative Region'],
    'Iran': ['Iran', 'Iran (Islamic Rep. of)', 'Iran (Islamic Republic of)', 'Iran, Islamic Rep.'],
    'Korea Dem. Peoples Rep.': ["Korea Dem. Peoples Rep.", "Democratic People's Republic of Korea", "Korea, Dem.Ppl's.Rep.", 'Korea, Dem. People’s Rep.'],
    'South Korea': ['South Korea', 'Korea', 'Korea, Republic of', 'Republic of Korea'],
    'Lao PDR': ['Lao PDR', "Lao People's Dem. Rep.", "Lao People's Democratic Republic"],
    'Macao SAR': ['Macao SAR', 'China, Macao SAR', 'Macao Special Administrative Region', 'China, Macao Special Administrative Region'],
    'Micronesia': ['Micronesia', 'Micronesia (Fed. States of)', 'Micronesia (Federated States of)', 'Micronesia, Federated States of'],
    'Moldova': ['Moldova', 'Republic of Moldova'],
    'North Macedonia': ['North Macedonia', 'Republic of North Macedonia', 'The former Yugoslav Republic of Macedonia', 'The Former Yugoslav Rep. of Macedonia'],
    'Russia': ['Russia', 'Russian Federation'],
    'Slovak Republic': ['Slovak Republic', 'Slovakia'],
    'Syria': ['Syria', 'Syrian Arab Republic'],
    'Tanzania': ['Tanzania', 'United Rep. of Tanzania', 'United Republic of Tanzania', 'United Republic of Tanzania: Mainland', 'United Republic of Tanzania: Zanzibar'],
    'Turkey': ['Turkey', 'Turkiye', 'Türkiye'],
    'United Kingdom': ['United Kingdom', 'United Kingdom of Great Britain and Northern Ireland', 'United Kingdom of Great Britan and Northern Ireland'],
    'United States': ['United States', 'United States of America', 'USA'],
    'Venezuela': ['Venezuela', 'Venezuela (Bolivarian Republic of)', 'Venezuela, RB'],
    'Vietnam': ['Vietnam', 'Viet Nam'],
    'West Bank and Gaza': ['West Bank and Gaza', 'Palestine', 'State of Palestine']
}

country_lookup = {}

for country_val in country_cols:
    country_lookup[norm_country_name(country_val)] = country_val

for canonical_val, alias_vals in merge_groups.items():
    if canonical_val in country_cols:
        for alias_val in alias_vals:
            country_lookup[norm_country_name(alias_val)] = canonical_val


def country_to_template(value_obj):
    norm_val = norm_country_name(value_obj)

    if norm_val in country_lookup:
        return country_lookup[norm_val]

    norm_val = re.sub(r'^(the )', '', norm_val).strip()

    if norm_val in country_lookup:
        return country_lookup[norm_val]

    return None


print('Country aliases loaded:')
print(len(country_lookup))

````

````python



# ============================================================
# 5. Create robust web session
# ============================================================

session_obj = requests.Session()

retry_obj = Retry(
    total=5,
    connect=5,
    read=5,
    backoff_factor=1.5,
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=['GET']
)

session_obj.mount('https://', HTTPAdapter(max_retries=retry_obj))
session_obj.mount('http://', HTTPAdapter(max_retries=retry_obj))

````

````python

# ============================================================
# 6. Scrape UNSD Excel links
# ============================================================

response_obj = session_obj.get(BASE_PAGE_URL, timeout=TIMEOUT)
response_obj.raise_for_status()

soup_obj = BeautifulSoup(response_obj.text, 'html.parser')

link_records = []

for anchor_obj in soup_obj.find_all('a'):
    href_val = anchor_obj.get('href')

    if href_val is None:
        continue

    href_lower = href_val.lower()

    if '.xls' not in href_lower:
        continue

    full_url = urljoin(BASE_PAGE_URL, href_val)

    indicator_text = clean_text(anchor_obj.get_text())

    if indicator_text == '':
        parent_text = clean_text(anchor_obj.parent.get_text()) if anchor_obj.parent is not None else ''
        indicator_text = parent_text

    if indicator_text == '':
        indicator_text = Path(urlparse(full_url).path).stem

    link_records.append({
        'indicator': indicator_text,
        'url': full_url
    })

links_df = pd.DataFrame(link_records).drop_duplicates('url').reset_index(drop=True)

print(links_df.head())
print('Excel links found:')
print(links_df.shape[0])

if links_df.shape[0] == 0:
    raise ValueError('No Excel links were found. The UNSD page layout or URL may have changed.')

````

````python

# ============================================================
# 7. Download UNSD Excel files
# ============================================================

download_records = []
failed_downloads = []

for record_obj in tqdm(links_df.to_dict('records'), desc='Downloading UNSD workbooks'):
    file_name = safe_filename(Path(urlparse(record_obj['url']).path).name)

    if file_name == '' or '.xls' not in file_name.lower():
        file_name = safe_filename(record_obj['indicator']) + '.xlsx'

    file_path = DOWNLOAD_DIR / file_name

    if not file_path.exists() or file_path.stat().st_size == 0:
        try:
            file_response = session_obj.get(record_obj['url'], timeout=(20, 120))
            file_response.raise_for_status()
            file_path.write_bytes(file_response.content)
            time.sleep(0.2)

        except Exception as exc_obj:
            failed_downloads.append({
                'indicator': record_obj['indicator'],
                'url': record_obj['url'],
                'error': str(exc_obj)[:300]
            })

    if file_path.exists() and file_path.stat().st_size > 0:
        download_records.append({
            'indicator': record_obj['indicator'],
            'url': record_obj['url'],
            'file_path': str(file_path),
            'file_name': file_name
        })

downloads_df = pd.DataFrame(download_records)
failed_downloads_df = pd.DataFrame(failed_downloads)

print(downloads_df.head())
print('Downloaded files:')
print(downloads_df.shape[0])

print('Failed downloads:')
print(failed_downloads_df.shape[0])

if failed_downloads_df.shape[0] > 0:
    print(failed_downloads_df.head(20))

````

````python
# ============================================================
# 8. Excel metadata parser
#     Read indicator title, definition, source notes, and unit
#     from the workbook description sheet.
# ============================================================

def get_description_metadata(file_path_obj, fallback_indicator):
    excel_obj = pd.ExcelFile(file_path_obj)

    # --------------------------------------------------------
    # 8A. Find the description sheet
    # --------------------------------------------------------

    desc_sheet = None

    for sheet_val in excel_obj.sheet_names:
        sheet_text = str(sheet_val).lower().strip()

        if sheet_text.startswith('desc'):
            desc_sheet = sheet_val
            break

    if desc_sheet is None:
        desc_sheet = excel_obj.sheet_names[0]


    # --------------------------------------------------------
    # 8B. Read description sheet without headers
    # --------------------------------------------------------

    desc_df = pd.read_excel(
        file_path_obj,
        sheet_name=desc_sheet,
        header=None,
        nrows=200
    )


    # --------------------------------------------------------
    # 8C. Helper to read a metadata value after a label cell
    # --------------------------------------------------------
    # This handles both structures:
    #
    #   Unit:        Number
    #
    # and:
    #
    #   Unit: Number
    #
    # It searches the description sheet cell by cell.

    def extract_value_after_label(label_text):
        label_lower = label_text.lower().strip().rstrip(':')

        for row_idx in range(desc_df.shape[0]):
            for col_idx in range(desc_df.shape[1]):
                cell_val = desc_df.iat[row_idx, col_idx]

                if not is_reported_value(cell_val):
                    continue

                cell_text = clean_text(cell_val)
                cell_lower = cell_text.lower().strip()

                # Case 1: label and value are in the same cell
                # Example: "Unit: Number"
                if cell_lower.startswith(label_lower + ':'):
                    same_cell_value = clean_text(
                        cell_text.split(':', 1)[1]
                    )

                    if same_cell_value != '':
                        return same_cell_value

                # Case 2: label is in one cell and value is to the right
                # Example: "Unit:" | "Number"
                if cell_lower.rstrip(':') == label_lower:
                    right_values = []

                    for next_col_idx in range(col_idx + 1, desc_df.shape[1]):
                        next_val = desc_df.iat[row_idx, next_col_idx]

                        if is_reported_value(next_val):
                            right_values.append(clean_text(next_val))

                    if len(right_values) > 0:
                        return clean_text(' '.join(right_values))

                    # Case 3: value is directly below the label
                    # This is a fallback for unusual layouts.
                    for next_row_idx in range(row_idx + 1, min(row_idx + 4, desc_df.shape[0])):
                        below_val = desc_df.iat[next_row_idx, col_idx]

                        if is_reported_value(below_val):
                            return clean_text(below_val)

        return ''


    # --------------------------------------------------------
    # 8D. Convert non-empty rows to row-level text
    # --------------------------------------------------------

    non_empty_rows = []

    for row_idx in range(desc_df.shape[0]):
        row_vals = []

        for col_idx in range(desc_df.shape[1]):
            val_obj = desc_df.iat[row_idx, col_idx]

            if is_reported_value(val_obj):
                row_vals.append(clean_text(val_obj))

        non_empty_rows.append(row_vals)


    # --------------------------------------------------------
    # 8E. Extract title
    # --------------------------------------------------------

    title_val = fallback_indicator

    for row_vals in non_empty_rows:
        if len(row_vals) == 0:
            continue

        joined_val = clean_text(' '.join(row_vals))
        lower_val = joined_val.lower()

        if joined_val == '':
            continue

        if 'united nations statistics division' in lower_val:
            continue

        if lower_val.startswith('date of release:'):
            continue

        if lower_val.startswith('coverage:'):
            continue

        if lower_val.startswith('series type:'):
            continue

        if lower_val.startswith('unit:'):
            continue

        if lower_val.startswith('definitions'):
            continue

        if lower_val.startswith('data quality'):
            continue

        if lower_val.startswith('source:'):
            continue

        title_val = joined_val
        break


    # --------------------------------------------------------
    # 8F. Extract unit directly from the description sheet
    # --------------------------------------------------------

    unit_val = extract_value_after_label('Unit')


    # --------------------------------------------------------
    # 8G. Extract definition and source notes
    # --------------------------------------------------------

    definition_parts = []
    source_parts = []
    mode_val = None

    for row_vals in non_empty_rows:
        joined_val = clean_text(' '.join(row_vals))
        lower_val = joined_val.lower()

        if joined_val == '':
            continue

        if lower_val.startswith('definitions'):
            mode_val = 'definition'
            continue

        if lower_val.startswith('data quality'):
            mode_val = None
            continue

        if lower_val.startswith('source:'):
            mode_val = 'source'

            source_text = clean_text(joined_val.split(':', 1)[1]) if ':' in joined_val else ''

            if source_text != '':
                source_parts.append(source_text)

            continue

        if lower_val.startswith('related information'):
            mode_val = None
            continue

        if lower_val.startswith('contact:'):
            mode_val = None
            continue

        if mode_val == 'definition':
            definition_parts.append(joined_val)

        if mode_val == 'source':
            source_parts.append(joined_val)

    definition_val = clean_text(' '.join(definition_parts))
    source_notes_val = clean_text(' '.join(source_parts))


    # --------------------------------------------------------
    # 8H. Return metadata
    # --------------------------------------------------------

    return {
        'indicator_title': title_val,
        'Definition': definition_val,
        'unit': unit_val,
        'source_notes': source_notes_val
    }
````

````python
# ============================================================
# 9. Excel data-sheet parser helpers
# ============================================================

def find_data_sheet(excel_obj):
    for sheet_val in excel_obj.sheet_names:
        if str(sheet_val).lower().strip() == 'data':
            return sheet_val

    for sheet_val in excel_obj.sheet_names:
        if 'data' in str(sheet_val).lower().strip():
            return sheet_val

    return excel_obj.sheet_names[-1]


def make_unique_headers(raw_headers):
    output_vals = []
    counts_obj = {}

    for raw_val in raw_headers:
        header_val = normalize_header(raw_val)

        if header_val == '' or header_val.lower() == '':
            header_val = 'unnamed'

        if header_val in counts_obj:
            counts_obj[header_val] += 1
            header_val = header_val + '_' + str(counts_obj[header_val])
        else:
            counts_obj[header_val] = 0

        output_vals.append(header_val)

    return output_vals


def find_header_row(raw_df):
    best_row = 0
    best_score = -1

    max_rows = min(20, raw_df.shape[0])

    for row_idx in range(max_rows):
        vals = [clean_text(x).lower() for x in raw_df.iloc[row_idx].tolist()]

        has_countryid = any(v == 'countryid' for v in vals)
        has_country = any(
            v in ['country', 'country and area', 'countries or areas']
            for v in vals
        )

        if has_countryid and has_country:
            return row_idx

        if any('latest year available' in v for v in vals) and row_idx <= 2:
            return row_idx

        score = 0

        if has_countryid:
            score += 100

        if has_country:
            score += 100

        score += sum(
            1 for v in vals
            if re.fullmatch(r'\d{4}|\d{4}-\d{4}', v)
        )

        if score > best_score:
            best_score = score
            best_row = row_idx

    return best_row


def make_disaster_headers(raw_df, header_idx):
    main_vals = raw_df.iloc[max(0, header_idx - 1)].tolist()
    sub_vals = raw_df.iloc[header_idx].tolist()

    headers = []
    current_group = ''

    for idx_val, sub_val in enumerate(sub_vals):
        main_val = normalize_header(main_vals[idx_val]) if idx_val < len(main_vals) else ''
        sub_text = normalize_header(sub_val)

        if main_val != '' and main_val.lower() != '':
            current_group = main_val

        if sub_text.lower() in ['countryid', 'country', 'country and area', 'countries or areas']:
            headers.append(sub_text)

        elif re.fullmatch(r'\d{4}-\d{4}', sub_text) and current_group != '':
            headers.append(current_group + ' | ' + sub_text)

        else:
            headers.append(sub_text)

    return make_unique_headers(headers)


def load_data_sheet(file_path_obj):
    excel_obj = pd.ExcelFile(file_path_obj)
    sheet_name_val = find_data_sheet(excel_obj)

    raw_df = pd.read_excel(
        file_path_obj,
        sheet_name=sheet_name_val,
        header=None
    )

    raw_df = (
        raw_df
        .dropna(how='all')
        .dropna(axis=1, how='all')
        .reset_index(drop=True)
    )

    header_idx = find_header_row(raw_df)
    header_vals = raw_df.iloc[header_idx].tolist()

    # Some UNSD files have blank first columns in the header row
    # when "Latest year available" appears in the same row.
    if len(header_vals) >= 2 and blank_like(header_vals[0]) and blank_like(header_vals[1]):
        if any('latest year available' in clean_text(x).lower() for x in header_vals):
            header_vals[0] = 'CountryID'
            header_vals[1] = 'Country'

    # Disaster files often have grouped headers:
    # main header above, period header below.
    if header_idx > 0 and any(
        re.fullmatch(r'\d{4}-\d{4}', clean_text(x))
        for x in header_vals
    ):
        headers = make_disaster_headers(raw_df, header_idx)
    else:
        headers = make_unique_headers(header_vals)

    # The row immediately below the header may contain data-table units.
    # We use this only to determine where data starts.
    # We should not use this row as the final metadata unit.
    units_idx = header_idx + 1 if header_idx + 1 < raw_df.shape[0] else header_idx
    units_row = raw_df.iloc[units_idx].tolist()

    unit_pattern = (
        r'tonne|tonnes|percent|%|kg|m3|km|hectare|hectares|'
        r'number|inhabitant|capita|usd|joule|wh|celsius|mm|mio'
    )

    unit_like = 0

    for x in units_row:
        if blank_like(x):
            unit_like += 1
        elif re.search(unit_pattern, clean_text(x).lower()):
            unit_like += 1

    if unit_like >= 3:
        data_start = header_idx + 2
    else:
        data_start = header_idx + 1

    data_df = raw_df.iloc[data_start:].copy().reset_index(drop=True)
    data_df.columns = headers

    return data_df, units_row, sheet_name_val


def find_country_column(data_df):
    for col_val in data_df.columns:
        low_val = str(col_val).lower().strip()

        if low_val in ['country', 'country and area', 'countries or areas']:
            return col_val

        if low_val.startswith('country '):
            return col_val

    for col_val in data_df.columns:
        low_val = str(col_val).lower().strip()

        if 'country' in low_val or 'countries' in low_val:
            return col_val

    return None


def clean_variable_name(col_val):
    text_val = normalize_header(col_val)
    text_val = re.sub(r'_\d+$', '', text_val)
    return clean_text(text_val)


def unit_for_column(col_idx, units_row, default_unit):
    # IMPORTANT:
    # default_unit should come from the description sheet metadata:
    # the cell labeled "Unit:".
    #
    # The data-sheet unit row is used only as a fallback.
    # This prevents values such as "number" from the data table
    # from replacing the official metadata unit.

    default_unit_clean = clean_text(default_unit)

    if default_unit_clean.lower() not in ['', 'none', 'nan', 'na', 'n/a']:
        return default_unit_clean

    # Fallback only if the description sheet has no Unit value.
    if col_idx < len(units_row):
        unit_val = clean_text(units_row[col_idx])

        if unit_val.lower() not in missing_tokens:
            return unit_val

    return ''


def period_rank(value_obj):
    text_val = clean_text(value_obj)
    found_vals = re.findall(r'\d{4}', text_val)

    if len(found_vals) == 0:
        return 9999

    return max([int(x) for x in found_vals])


def split_variable_period(col_val):
    text_val = clean_variable_name(col_val)

    if ' | ' in text_val:
        left_val, right_val = text_val.rsplit(' | ', 1)
        return clean_text(left_val), clean_text(right_val)

    return text_val, ''


id_cols_lower = set([
    'countryid',
    'country',
    'country and area',
    'countries or areas',
    'source',
    'city',
    'latest year available',
    'year'
])
````

````python
# ============================================================
# 10. Workbook parser
# ============================================================

def reported_by_country(sub_df, country_col, value_col):
    availability = {country_val: 0 for country_val in country_cols}

    for _, row_obj in sub_df.iterrows():
        template_country = country_to_template(row_obj.get(country_col))

        if template_country is None:
            continue

        if is_reported_value(row_obj.get(value_col)):
            availability[template_country] = 1

    return availability


def parse_workbook(record_obj):
    file_path_obj = Path(record_obj['file_path'])

    # Read workbook-level metadata from the description sheet.
    # The unit should come from the description sheet cell labeled "Unit:".
    meta_obj = get_description_metadata(file_path_obj, record_obj['indicator'])

    # Read the data sheet.
    # units_row is kept only because load_data_sheet returns it.
    # It should not replace the metadata unit in the final output.
    data_df, units_row, sheet_name_val = load_data_sheet(file_path_obj)

    country_col = find_country_column(data_df)

    if country_col is None:
        raise ValueError('Country column not found')

    rows_out = []

    metadata_unit = clean_text(meta_obj.get('unit', ''))


    # --------------------------------------------------------
    # 10A. Annual datasets with year columns
    # --------------------------------------------------------
    # Logic:
    # availability = 1 if a country has at least one reported value
    # in either of the two most recent years.

    year_cols = [
        col_val for col_val in data_df.columns
        if is_year_label(col_val)
    ]

    if len(year_cols) >= 2:
        year_vals = sorted([
            int(clean_text(col_val))
            for col_val in year_cols
        ])

        recent_years = [
            str(year_vals[-2]),
            str(year_vals[-1])
        ]

        availability = {country_val: 0 for country_val in country_cols}

        for _, row_obj in data_df.iterrows():
            template_country = country_to_template(row_obj.get(country_col))

            if template_country is None:
                continue

            has_value = False

            for year_col in recent_years:
                if is_reported_value(row_obj.get(year_col)):
                    has_value = True
                    break

            if has_value:
                availability[template_country] = 1

        row_dict = {
            'indicator': meta_obj['indicator_title'],
            'Definition': meta_obj['Definition'],
            'unit': metadata_unit,
            'source': record_obj['url'],
            '_file_name': record_obj['file_name'],
            '_url': record_obj['url'],
            '_period': max(year_vals),
            '_sheet': sheet_name_val,
            '_variable': meta_obj['indicator_title']
        }

        row_dict.update(availability)
        rows_out.append(row_dict)

        return rows_out


    # --------------------------------------------------------
    # 10B. Datasets with an explicit Year column
    # --------------------------------------------------------
    # Logic:
    # keep only rows from the latest available year before checking
    # availability.

    year_col = None

    for col_val in data_df.columns:
        if str(col_val).lower().strip() == 'year':
            year_col = col_val
            break

    if year_col is not None:
        year_numeric = pd.to_numeric(data_df[year_col], errors='coerce')

        if year_numeric.notna().any():
            latest_period = int(year_numeric.max())
            data_df = data_df.loc[year_numeric == latest_period].copy()
        else:
            latest_period = 9999
    else:
        latest_period = 9999


    # --------------------------------------------------------
    # 10C. Group non-ID variables
    # --------------------------------------------------------
    # For files with multiple variables or period columns, this groups
    # related columns and keeps the most recent period within each group.

    variable_groups = {}

    for idx_val, col_val in enumerate(data_df.columns):
        low_val = str(col_val).lower().strip()

        if low_val in id_cols_lower:
            continue

        if 'footnote' in low_val:
            continue

        if col_val == country_col:
            continue

        variable_name, period_label = split_variable_period(col_val)

        if variable_name.lower() in missing_tokens:
            continue

        if variable_name.lower() == 'unnamed':
            continue

        if variable_name not in variable_groups:
            variable_groups[variable_name] = []

        variable_groups[variable_name].append(
            (
                idx_val,
                col_val,
                period_label,
                period_rank(period_label)
            )
        )


    # --------------------------------------------------------
    # 10D. Create one output row per variable group
    # --------------------------------------------------------

    for variable_name, group_vals in variable_groups.items():
        best_rank = max([x[3] for x in group_vals])
        best_cols = [x for x in group_vals if x[3] == best_rank]

        availability = {country_val: 0 for country_val in country_cols}

        for _, row_obj in data_df.iterrows():
            template_country = country_to_template(row_obj.get(country_col))

            if template_country is None:
                continue

            has_value = False

            for _, col_val, _, _ in best_cols:
                if is_reported_value(row_obj.get(col_val)):
                    has_value = True
                    break

            if has_value:
                availability[template_country] = 1

        if len(variable_groups) == 1:
            indicator_name = meta_obj['indicator_title']
        else:
            indicator_name = variable_name

        row_dict = {
            'indicator': indicator_name,
            'Definition': meta_obj['Definition'],
            'unit': metadata_unit,
            'source': record_obj['url'],
            '_file_name': record_obj['file_name'],
            '_url': record_obj['url'],
            '_period': best_rank if best_rank != 9999 else latest_period,
            '_sheet': sheet_name_val,
            '_variable': variable_name
        }

        row_dict.update(availability)
        rows_out.append(row_dict)

    return rows_out
````

````python
# ============================================================
# 11. Parse all downloaded workbooks
# ============================================================

output_rows = []
parse_errors = []

for record_obj in tqdm(downloads_df.to_dict('records'), desc='Parsing UNSD workbooks'):
    try:
        workbook_rows = parse_workbook(record_obj)

        if workbook_rows is not None and len(workbook_rows) > 0:
            output_rows.extend(workbook_rows)

    except Exception as exc_obj:
        parse_errors.append({
            'file_name': record_obj.get('file_name'),
            'indicator': record_obj.get('indicator'),
            'error': str(exc_obj)[:300]
        })


# ------------------------------------------------------------
# 11A. Convert parser outputs to DataFrames
# ------------------------------------------------------------

parsed_df = pd.DataFrame(output_rows)
parse_errors_df = pd.DataFrame(parse_errors)


# ------------------------------------------------------------
# 11B. Basic parser summary
# ------------------------------------------------------------

print('Parsed shape:')
print(parsed_df.shape)

print('\nParse errors:')
print(parse_errors_df.shape[0])

if parse_errors_df.shape[0] > 0:
    print('\nParse error examples:')
    display(parse_errors_df.head(20))


# ------------------------------------------------------------
# 11C. Show parsed output preview
# ------------------------------------------------------------

preview_cols = [
    'indicator',
    'unit',
    '_file_name',
    '_period',
    '_sheet',
    '_variable'
]

available_preview_cols = [
    col for col in preview_cols
    if col in parsed_df.columns
]

if parsed_df.shape[0] > 0 and len(available_preview_cols) > 0:
    print('\nParsed output preview:')
    display(parsed_df[available_preview_cols].head(20))
else:
    print('\nNo parsed rows available for preview.')


# ------------------------------------------------------------
# 11D. Check unit extraction from description metadata
# ------------------------------------------------------------

if parsed_df.shape[0] > 0 and 'unit' in parsed_df.columns:
    print('\nUnit value counts:')
    display(
        parsed_df['unit']
        .fillna('')
        .astype(str)
        .str.strip()
        .replace('', '[blank]')
        .value_counts()
        .head(30)
    )

    print('\nRows with blank unit:')
    blank_unit_df = parsed_df[
        parsed_df['unit']
        .fillna('')
        .astype(str)
        .str.strip()
        .eq('')
    ].copy()

    print(blank_unit_df.shape[0])

    if blank_unit_df.shape[0] > 0:
        blank_preview_cols = [
            col for col in ['indicator', 'unit', '_file_name', '_sheet', '_period']
            if col in blank_unit_df.columns
        ]

        display(blank_unit_df[blank_preview_cols].head(30))


# ------------------------------------------------------------
# 11E. Specific check for climatological disaster files
# ------------------------------------------------------------

if parsed_df.shape[0] > 0 and '_file_name' in parsed_df.columns:
    climatological_check_df = parsed_df[
        parsed_df['_file_name']
        .astype(str)
        .str.lower()
        .str.contains('climatological', na=False)
    ].copy()

    print('\nClimatological disaster unit check:')
    print(climatological_check_df.shape[0])

    if climatological_check_df.shape[0] > 0:
        clim_preview_cols = [
            col for col in ['indicator', 'unit', '_file_name', '_sheet', '_period', '_variable']
            if col in climatological_check_df.columns
        ]

        display(
            climatological_check_df[clim_preview_cols]
            .drop_duplicates()
            .head(50)
        )


# ------------------------------------------------------------
# 11F. Stop early if no rows were parsed
# ------------------------------------------------------------

if parsed_df.shape[0] == 0:
    raise ValueError(
        'No parsed rows were produced. Check parse_errors_df and the downloaded workbook files.'
    )
````

````python
# ============================================================
# 12. Final cleanup
#     - remove blank indicator rows
#     - preserve metadata unit from description sheet
#     - create helper columns for deduplication
#     - remove older duplicate year rows
#     - build final output using the template column structure
# ============================================================

if parsed_df.shape[0] == 0:
    raise ValueError('No parsed rows found. Check downloads and parser errors.')

parsed_df = parsed_df.copy()


# ------------------------------------------------------------
# 12A. Check required parser columns
# ------------------------------------------------------------

minimum_parser_cols = ['indicator', 'Definition', 'unit', 'source']

missing_parser_cols = [
    col for col in minimum_parser_cols
    if col not in parsed_df.columns
]

if len(missing_parser_cols) > 0:
    raise KeyError(
        'The following required parser columns are missing from parsed_df: '
        + ', '.join(missing_parser_cols)
    )


# ------------------------------------------------------------
# 12B. Clean core metadata columns
# ------------------------------------------------------------

for text_col in ['indicator', 'Definition', 'unit', 'source']:
    parsed_df[text_col] = (
        parsed_df[text_col]
        .fillna('')
        .astype(str)
        .str.replace('\xa0', ' ', regex=False)
        .str.replace(r'\s+', ' ', regex=True)
        .str.strip()
    )


# ------------------------------------------------------------
# 12C. Remove blank indicator rows
# ------------------------------------------------------------

parsed_df = parsed_df[
    ~parsed_df['indicator'].str.lower().isin(['', 'none', 'nan', 'na', 'n/a'])
].copy()


# ------------------------------------------------------------
# 12D. Helper function to remove trailing year from indicator
# ------------------------------------------------------------
# Example:
# Forest Area_c_Forest Area, 1990 -> Forest Area_c_Forest Area
# Forest Area_c_Forest Area, 2020 -> Forest Area_c_Forest Area

def get_base_indicator(indicator_value):
    text_val = clean_text(indicator_value)
    text_val = re.sub(r'\s*,\s*((18|19|20|21)\d{2})\s*$', '', text_val)
    return text_val.strip()


# ------------------------------------------------------------
# 12E. Helper function to extract trailing year from indicator
# ------------------------------------------------------------
# Example:
# Forest Area_c_Forest Area, 2020 -> 2020

def get_indicator_year(indicator_value):
    text_val = clean_text(indicator_value)
    match_obj = re.search(r'\s*,\s*((18|19|20|21)\d{2})\s*$', text_val)

    if match_obj:
        return int(match_obj.group(1))

    return np.nan


# ------------------------------------------------------------
# 12F. Create helper columns for deduplication
# ------------------------------------------------------------

parsed_df['_base_indicator_for_dedup'] = parsed_df['indicator'].apply(get_base_indicator)
parsed_df['_indicator_year_for_dedup'] = parsed_df['indicator'].apply(get_indicator_year)

print('Deduplication helper columns created:')

dedup_preview_cols = [
    col for col in [
        'indicator',
        'unit',
        '_file_name',
        '_period',
        '_base_indicator_for_dedup',
        '_indicator_year_for_dedup'
    ]
    if col in parsed_df.columns
]

display(parsed_df[dedup_preview_cols].head(10))


# ------------------------------------------------------------
# 12G. Split rows into year-based rows and non-year rows
# ------------------------------------------------------------

year_rows_df = parsed_df[
    parsed_df['_indicator_year_for_dedup'].notna()
].copy()

non_year_rows_df = parsed_df[
    parsed_df['_indicator_year_for_dedup'].isna()
].copy()

print('Rows with trailing indicator year:')
print(year_rows_df.shape[0])

print('Rows without trailing indicator year:')
print(non_year_rows_df.shape[0])


# ------------------------------------------------------------
# 12H. For duplicate year rows, keep only the most recent year
# ------------------------------------------------------------
# Duplicate definition:
# same source file + same base indicator
#
# Example:
# Forest Area_c_Forest Area, 1990
# Forest Area_c_Forest Area, 2000
# Forest Area_c_Forest Area, 2010
# Forest Area_c_Forest Area, 2015
# Forest Area_c_Forest Area, 2020
#
# becomes:
# Forest Area_c_Forest Area, 2020

if year_rows_df.shape[0] > 0:
    sort_cols_year = [
        col for col in [
            '_file_name',
            '_base_indicator_for_dedup',
            '_indicator_year_for_dedup'
        ]
        if col in year_rows_df.columns
    ]

    if len(sort_cols_year) > 0:
        year_rows_df = year_rows_df.sort_values(sort_cols_year)

    dedup_cols_year = [
        col for col in [
            '_file_name',
            '_base_indicator_for_dedup'
        ]
        if col in year_rows_df.columns
    ]

    if len(dedup_cols_year) == 0:
        dedup_cols_year = ['indicator']

    year_rows_df = (
        year_rows_df
        .drop_duplicates(dedup_cols_year, keep='last')
        .copy()
    )


# ------------------------------------------------------------
# 12I. For non-year rows, remove duplicate indicator rows
# ------------------------------------------------------------

if non_year_rows_df.shape[0] > 0:
    dedup_cols_non_year = []

    if '_file_name' in non_year_rows_df.columns:
        dedup_cols_non_year.append('_file_name')

    dedup_cols_non_year.append('indicator')

    sort_cols_non_year = []

    if '_file_name' in non_year_rows_df.columns:
        sort_cols_non_year.append('_file_name')

    if '_variable' in non_year_rows_df.columns:
        sort_cols_non_year.append('_variable')

    if '_period' in non_year_rows_df.columns:
        sort_cols_non_year.append('_period')

    if len(sort_cols_non_year) > 0:
        non_year_rows_df = non_year_rows_df.sort_values(sort_cols_non_year)

    non_year_rows_df = (
        non_year_rows_df
        .drop_duplicates(dedup_cols_non_year, keep='last')
        .copy()
    )


# ------------------------------------------------------------
# 12J. Combine cleaned rows back together
# ------------------------------------------------------------

parsed_df = pd.concat(
    [non_year_rows_df, year_rows_df],
    ignore_index=True
)

sort_cols_final = []

if '_file_name' in parsed_df.columns:
    sort_cols_final.append('_file_name')

if '_base_indicator_for_dedup' in parsed_df.columns:
    sort_cols_final.append('_base_indicator_for_dedup')

if '_indicator_year_for_dedup' in parsed_df.columns:
    sort_cols_final.append('_indicator_year_for_dedup')

if len(sort_cols_final) > 0:
    parsed_df = parsed_df.sort_values(
        sort_cols_final,
        na_position='last'
    ).reset_index(drop=True)


# ------------------------------------------------------------
# 12K. Check required output columns
# ------------------------------------------------------------

required_output_cols = ['indicator', 'Definition', 'unit', 'source'] + country_cols

missing_output_cols = [
    col for col in required_output_cols
    if col not in parsed_df.columns
]

if len(missing_output_cols) > 0:
    raise KeyError(
        'The following required columns are missing from parsed_df: '
        + ', '.join(missing_output_cols)
    )


# ------------------------------------------------------------
# 12L. Build final output using template column structure
# ------------------------------------------------------------

final_df = parsed_df[required_output_cols].copy()

final_df.insert(0, 'no', range(1, len(final_df) + 1))


# ------------------------------------------------------------
# 12M. Final Step 12 diagnostics
# ------------------------------------------------------------

print('Final output preview:')
display(final_df.head())

print('Final output shape:')
print(final_df.shape)

print('Unit value counts in final output:')
display(
    final_df['unit']
    .fillna('')
    .astype(str)
    .str.strip()
    .replace('', '[blank]')
    .value_counts()
    .head(30)
)
````

````python
 # ============================================================
# 13. Export to Excel
# ============================================================

# Export final DataFrame
final_df.to_excel(OUTPUT_XLSX, index=False)

# Load workbook for formatting
wb_obj = load_workbook(OUTPUT_XLSX)
ws_obj = wb_obj.active
ws_obj.title = 'Availability'


# ------------------------------------------------------------
# 13A. Add hyperlinks to the source column
# ------------------------------------------------------------

# Column positions:
# no = 1
# indicator = 2
# Definition = 3
# unit = 4
# source = 5

source_col_idx = 5

for row_idx in range(2, ws_obj.max_row + 1):
    cell_obj = ws_obj.cell(row=row_idx, column=source_col_idx)

    if cell_obj.value:
        cell_obj.hyperlink = str(cell_obj.value)
        cell_obj.style = 'Hyperlink'


# ------------------------------------------------------------
# 13B. Format header row
# ------------------------------------------------------------

for col_idx in range(1, ws_obj.max_column + 1):
    header_cell = ws_obj.cell(row=1, column=col_idx)
    header_cell.font = Font(bold=True)
    header_cell.alignment = Alignment(
        wrap_text=True,
        horizontal='center',
        vertical='center'
    )


# ------------------------------------------------------------
# 13C. Format all cells
# ------------------------------------------------------------

for row in ws_obj.iter_rows():
    for cell in row:
        cell.alignment = Alignment(
            wrap_text=True,
            vertical='top'
        )


# ------------------------------------------------------------
# 13D. Adjust column widths
# ------------------------------------------------------------

for col_idx in range(1, ws_obj.max_column + 1):
    col_letter = get_column_letter(col_idx)

    if col_idx == 1:
        ws_obj.column_dimensions[col_letter].width = 8

    elif col_idx == 2:
        ws_obj.column_dimensions[col_letter].width = 40

    elif col_idx == 3:
        ws_obj.column_dimensions[col_letter].width = 60

    elif col_idx == 4:
        ws_obj.column_dimensions[col_letter].width = 25

    elif col_idx == 5:
        ws_obj.column_dimensions[col_letter].width = 45

    else:
        ws_obj.column_dimensions[col_letter].width = 16


# ------------------------------------------------------------
# 13E. Freeze panes and add autofilter
# ------------------------------------------------------------

ws_obj.freeze_panes = 'F2'
ws_obj.auto_filter.ref = ws_obj.dimensions


# ------------------------------------------------------------
# 13F. Save workbook
# ------------------------------------------------------------

wb_obj.save(OUTPUT_XLSX)

print('Saved final workbook:')
print(OUTPUT_XLSX)
````

````python
 # ============================================================
# 14. Final quality checks
# ============================================================

print('============================================================')
print('FINAL QUALITY CHECKS')
print('============================================================')

print('Final output shape:')
print(final_df.shape)

print('\nNumber of indicator rows:')
print(final_df.shape[0])

print('\nNumber of country columns:')
print(len(country_cols))

print('\nFailed downloads:')
print(failed_downloads_df.shape[0])

if failed_downloads_df.shape[0] > 0:
    print('\nFailed download examples:')
    display(failed_downloads_df.head(20))

print('\nParse errors:')
print(parse_errors_df.shape[0])

if parse_errors_df.shape[0] > 0:
    print('\nParse error examples:')
    display(parse_errors_df.head(20))


# ------------------------------------------------------------
# 14A. Availability summary per indicator
# ------------------------------------------------------------

country_availability_counts = final_df[country_cols].sum(axis=1)

print('\nAvailability summary per indicator:')
print(country_availability_counts.describe())


# ------------------------------------------------------------
# 14B. Indicators with zero country availability
# ------------------------------------------------------------

zero_availability_df = final_df.loc[
    country_availability_counts == 0,
    ['no', 'indicator', 'unit', 'source']
].copy()

print('\nIndicators with zero country availability:')
print(zero_availability_df.shape[0])

if zero_availability_df.shape[0] > 0:
    display(zero_availability_df.head(20))


# ------------------------------------------------------------
# 14C. Country-level availability summary
# ------------------------------------------------------------

country_summary_df = pd.DataFrame({
    'country': country_cols,
    'available_indicators': final_df[country_cols].sum(axis=0).values
})

country_summary_df['total_indicators'] = final_df.shape[0]
country_summary_df['availability_rate_percent'] = (
    country_summary_df['available_indicators']
    / country_summary_df['total_indicators']
    * 100
).round(2)

country_summary_df = country_summary_df.sort_values(
    'available_indicators',
    ascending=False
).reset_index(drop=True)

print('\nCountry-level availability summary:')
display(country_summary_df.head(20))


# ------------------------------------------------------------
# 14D. Save country-level summary as a separate Excel file
# ------------------------------------------------------------

COUNTRY_SUMMARY_XLSX = 'UNSD_Country_Availability_Summary.xlsx'

country_summary_df.to_excel(COUNTRY_SUMMARY_XLSX, index=False)

print('\nSaved country-level summary:')
print(COUNTRY_SUMMARY_XLSX)


# ------------------------------------------------------------
# 14E. Save parse errors and failed downloads, if any
# ------------------------------------------------------------

if failed_downloads_df.shape[0] > 0:
    FAILED_DOWNLOADS_XLSX = 'UNSD_Failed_Downloads.xlsx'
    failed_downloads_df.to_excel(FAILED_DOWNLOADS_XLSX, index=False)

    print('\nSaved failed downloads file:')
    print(FAILED_DOWNLOADS_XLSX)

if parse_errors_df.shape[0] > 0:
    PARSE_ERRORS_XLSX = 'UNSD_Parse_Errors.xlsx'
    parse_errors_df.to_excel(PARSE_ERRORS_XLSX, index=False)

    print('\nSaved parse errors file:')
    print(PARSE_ERRORS_XLSX)


print('\nDone.')
````

````python
# ============================================================
# 1. Paths and settings
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

# Input files
WB_FINAL_PATH = "WB_final.csv"

# If you already downloaded the CWON file, use this local path.
# Otherwise, the code will download it from the World Bank URL.
CWON_LOCAL_PATH = "WB_CWON_WIDEF.csv"
CWON_URL = "https://data360files.worldbank.org/data360-data/data/WB_CWON/WB_CWON_WIDEF.csv"

# Output file
OUTPUT_PATH = "WB_final_CWON_full_indicator_variations.csv"

# Number of latest reporting years to check.
# Example: if the latest columns are 2019 and 2020, a country gets 1 if either year has data.
N_LATEST_YEARS = 2
````

````python
# ============================================================
# 3. Identify template country columns
# ============================================================

# The first five columns in WB_final are assumed to be:
# no, indicator, Definition, unit, source
base_columns = ["no", "indicator", "Definition", "unit", "source"]

# Country columns start after the first five columns
country_cols = list(wb_final.columns[5:])

print("Number of country columns in WB_final:")
print(len(country_cols))

print("First 10 country columns:")
print(country_cols[:10])
````

````python
# ============================================================
# 4. Identify latest reporting years in CWON
# ============================================================

# Year columns are columns whose names are all digits, e.g. 1995, 1996, ..., 2020
year_cols = [col for col in cwon.columns if str(col).isdigit()]
year_cols = sorted(year_cols, key=lambda x: int(x))

# Use the latest N years
latest_year_cols = year_cols[-N_LATEST_YEARS:]

print("All year columns:")
print(year_cols)

print("Latest reporting years used for availability:")
print(latest_year_cols)
````

````python
# ============================================================
# 5. Normalize country names to match WB_final
# ============================================================

# CWON country names may differ slightly from WB_final country names.
# This mapping converts CWON names into WB_final names.
country_name_map = {
    "Bahamas, The": "Bahamas",
    "Congo, Dem. Rep.": "Congo Dem. Rep.",
    "Congo, Rep.": "Congo Rep.",
    "Cote d'Ivoire": "Côte dIvoire",
    "Egypt, Arab Rep.": "Egypt",
    "Gambia, The": "Gambia",
    "Hong Kong SAR, China": "Hong Kong (S.A.R.)",
    "Iran, Islamic Rep.": "Iran",
    "Korea, Dem. People's Rep.": "Korea Dem. Peoples Rep.",
    "Korea, Rep.": "South Korea",
    "Macao SAR, China": "Macao SAR",
    "Micronesia, Fed. Sts.": "Micronesia",
    "Russian Federation": "Russia",
    "Sao Tome and Principe": "São Tomé and Principe",
    "Turkiye": "Turkey",
    "Venezuela, RB": "Venezuela",
    "Yemen, Rep.": "Yemen"
}

# Create a working copy of CWON
work_df = cwon.copy()

# Apply country-name mapping
work_df["template_country"] = work_df["REF_AREA_LABEL"].replace(country_name_map)

# Keep only countries that exist in WB_final
work_df = work_df[work_df["template_country"].isin(country_cols)].copy()

print("Number of CWON rows after matching countries to WB_final:")
print(work_df.shape[0])

print("Number of matched countries:")
print(work_df["template_country"].nunique())
````

````python
# ============================================================
# 6. Define the full CWON indicator variation
# ============================================================

# The revised indicator definition is the combination of:
# INDICATOR_LABEL + SEX_LABEL + COMP_BREAKDOWN_1_LABEL + COMP_BREAKDOWN_2_LABEL

indicator_component_cols = [
    "INDICATOR_LABEL",
    "SEX_LABEL",
    "COMP_BREAKDOWN_1_LABEL",
    "COMP_BREAKDOWN_2_LABEL"
]

metadata_cols = [
    "INDICATOR_LABEL",
    "SEX_LABEL",
    "COMP_BREAKDOWN_1_LABEL",
    "COMP_BREAKDOWN_2_LABEL",
    "UNIT_MEASURE_LABEL",
    "DATABASE_ID_LABEL"
]

# Fill missing metadata values so grouping and labels work cleanly
for col in metadata_cols:
    work_df[col] = work_df[col].fillna("Not Available")

# Create combined indicator name
work_df["combined_indicator"] = (
    work_df["INDICATOR_LABEL"] + " | " +
    work_df["SEX_LABEL"] + " | " +
    work_df["COMP_BREAKDOWN_1_LABEL"] + " | " +
    work_df["COMP_BREAKDOWN_2_LABEL"]
)

# Create definition field
work_df["definition_text"] = (
    "SEX_LABEL: " + work_df["SEX_LABEL"] + "; " +
    "COMP_BREAKDOWN_1_LABEL: " + work_df["COMP_BREAKDOWN_1_LABEL"] + "; " +
    "COMP_BREAKDOWN_2_LABEL: " + work_df["COMP_BREAKDOWN_2_LABEL"] + "; " +
    "DATABASE_ID_LABEL: " + work_df["DATABASE_ID_LABEL"]
)

print("Number of unique full indicator variations:")
print(work_df["combined_indicator"].nunique())

print("Example full indicators:")
print(work_df[["combined_indicator", "definition_text", "UNIT_MEASURE_LABEL", "DATABASE_ID_LABEL"]].drop_duplicates().head(10))
````

````python
# ============================================================
# 7. Calculate country availability for latest reporting years
# ============================================================

# Convert latest year columns to numeric
for col in latest_year_cols:
    work_df[col] = pd.to_numeric(work_df[col], errors="coerce")

# Availability equals 1 if at least one of the latest reporting years has data
work_df["available_latest"] = work_df[latest_year_cols].notna().any(axis=1).astype(int)

# Group by the full indicator variation and country
group_cols = [
    "combined_indicator",
    "definition_text",
    "UNIT_MEASURE_LABEL",
    "DATABASE_ID_LABEL",
    "template_country"
]

availability_long = (
    work_df
    .groupby(group_cols, dropna=False)["available_latest"]
    .max()
    .reset_index()
)

print("Long availability table shape:")
print(availability_long.shape)

print("Long availability preview:")
print(availability_long.head())
````

````python
# ============================================================
# 8. Pivot availability to country-wide format
# ============================================================

availability_pivot = availability_long.pivot_table(
    index=[
        "combined_indicator",
        "definition_text",
        "UNIT_MEASURE_LABEL",
        "DATABASE_ID_LABEL"
    ],
    columns="template_country",
    values="available_latest",
    aggfunc="max",
    fill_value=0
).reset_index()

# Make sure every country from WB_final exists as a column
for country in country_cols:
    if country not in availability_pivot.columns:
        availability_pivot[country] = 0

print("Wide availability table shape:")
print(availability_pivot.shape)

print("Wide availability preview:")
print(availability_pivot.head())
````

````python
# ============================================================
# 9. Fill WB_final template structure
# ============================================================

# Create final output dataframe using the same column structure as WB_final
final_df = pd.DataFrame()

final_df["no"] = range(1, len(availability_pivot) + 1)
final_df["indicator"] = availability_pivot["combined_indicator"]
final_df["Definition"] = availability_pivot["definition_text"]
final_df["unit"] = availability_pivot["UNIT_MEASURE_LABEL"]
final_df["source"] = availability_pivot["DATABASE_ID_LABEL"]

# Add country availability columns
for country in country_cols:
    final_df[country] = availability_pivot[country].astype(int)

# Reorder to exactly match WB_final columns
final_df = final_df[wb_final.columns]

print("Final output shape:")
print(final_df.shape)

print("Final output preview:")
print(final_df.head())
````

````python
# ============================================================
# 10. Validation checks
# ============================================================

# Check example requested by user: Angola availability for Domestic comprehensive wealth index
angola_check = final_df.loc[
    final_df["indicator"].str.contains("Domestic comprehensive wealth index", case=False, na=False),
    ["no", "indicator", "Definition", "unit", "source", "Angola"]
]

print("Angola check for Domestic comprehensive wealth index:")
print(angola_check)

# Check how many countries have availability for each indicator
country_availability_counts = final_df[country_cols].sum(axis=1)

summary_check = final_df[["no", "indicator", "unit", "source"]].copy()
summary_check["available_country_count"] = country_availability_counts

print("Availability count summary:")
print(summary_check.head(20))

print("Minimum available countries for an indicator:")
print(summary_check["available_country_count"].min())

print("Maximum available countries for an indicator:")
print(summary_check["available_country_count"].max())
````

````python
# ============================================================
# IMF Climate Data Scraping Script
# Purpose:
#   1. Download the IMF Climate indicators summary.
#   2. Filter to country-level datasets.
#   3. Download each country-level dataset as CSV.
#   4. Save a manifest describing which datasets were downloaded.
#
# Output folders:
#   imf_climate_raw_downloads
#   imf_climate_country_coverage_datasets
#
# Output files:
#   00_Indicators_Summary.csv
#   imf_climate_country_coverage_manifest.csv
# ============================================================
````

````python
 # ============================================================
# 1. Import packages
# ============================================================

import os
import re
import io
import sys
import time
import zipfile
import subprocess
import unicodedata
import requests
import pandas as pd
from tqdm.auto import tqdm
````

````python
 # ============================================================
# 2. Paths and settings
# ============================================================

# IMF / ArcGIS catalog item for the indicators summary
catalog_item_id = "74a9db2b716d46f2b25c2ec53c0ed1ce"

# Output folders
raw_download_folder = "imf_climate_raw_downloads"
dataset_folder = "imf_climate_country_coverage_datasets"

os.makedirs(raw_download_folder, exist_ok=True)
os.makedirs(dataset_folder, exist_ok=True)

# Output files
indicators_summary_file = "00_Indicators_Summary.csv"

manifest_file = os.path.join(
    dataset_folder,
    "imf_climate_country_coverage_manifest.csv"
)

# Request settings
request_timeout = 180
pause_seconds = 0.25

# Browser-like request headers
headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 Chrome/124.0 Safari/537.36"
    ),
    "Accept": "text/csv,application/csv,application/json,text/plain,*/*",
    "Referer": "https://climatedata.imf.org/",
    "Origin": "https://climatedata.imf.org"
}

print("Scraping paths and settings are ready.")
print("Raw download folder:", raw_download_folder)
print("Dataset folder:", dataset_folder)
print("Manifest file:", manifest_file)
````

````python
# ============================================================
# 3. Helper functions for scraping
# ============================================================

def clean_text(value):
    """Clean text values by removing extra spaces."""
    if pd.isna(value):
        return ""
    return re.sub(r"\s+", " ", str(value).strip())


def slugify(value, max_len=140):
    """
    Convert indicator names into safe file names.
    Example:
        'Carbon Tax to Revenues/Assets' -> 'Carbon_Tax_to_Revenues_Assets'
    """
    text_value = unicodedata.normalize("NFKD", str(value))
    text_value = text_value.encode("ascii", "ignore").decode("ascii")
    text_value = re.sub(r"[^A-Za-z0-9]+", "_", text_value)
    text_value = re.sub(r"_+", "_", text_value).strip("_")

    if text_value == "":
        text_value = "dataset"

    return text_value[:max_len]


def normalize_columns(df):
    """
    Clean column names and remove unnamed columns.
    """
    df = df.copy()

    df.columns = [
        clean_text(col)
        for col in df.columns
    ]

    unnamed_cols = [
        col
        for col in df.columns
        if str(col).lower().startswith("unnamed")
    ]

    if len(unnamed_cols) > 0:
        df = df.drop(columns=unnamed_cols)

    return df


def extract_item_id_from_url(url_value):
    """
    Extract ArcGIS item ID from IMF download URLs.
    """
    url_text = clean_text(url_value)

    match = re.search(r"/datasets/([A-Za-z0-9]+)/", url_text)
    if match is not None:
        return match.group(1)

    match = re.search(r"/items/([A-Za-z0-9]+)", url_text)
    if match is not None:
        return match.group(1)

    return ""


def get_url_content(url_value):
    """
    Download content from a URL.
    """
    response = requests.get(
        url_value,
        headers=headers,
        timeout=request_timeout
    )

    response.raise_for_status()
    return response.content


def read_csv_from_response_bytes(content_bytes):
    """
    Read CSV content from either direct CSV bytes or ZIP bytes.
    Some ArcGIS endpoints return ZIP files containing CSV files.
    """
    # ZIP files usually begin with b"PK"
    if content_bytes[:2] == b"PK":
        with zipfile.ZipFile(io.BytesIO(content_bytes)) as zip_file:
            csv_names = [
                name
                for name in zip_file.namelist()
                if name.lower().endswith(".csv")
            ]

            if len(csv_names) == 0:
                raise ValueError("ZIP response did not contain any CSV files.")

            csv_names = sorted(csv_names)

            with zip_file.open(csv_names[0]) as file_obj:
                df = pd.read_csv(file_obj, low_memory=False)

            return normalize_columns(df)

    # Direct CSV response
    df = pd.read_csv(io.BytesIO(content_bytes), low_memory=False)
    return normalize_columns(df)


def try_read_csv_from_urls(urls):
    """
    Try several possible download URLs.
    Return the first successful dataframe, URL, and raw content bytes.
    """
    last_error = None

    for url_value in urls:
        try:
            content_bytes = get_url_content(url_value)
            df = read_csv_from_response_bytes(content_bytes)
            return df, url_value, content_bytes

        except Exception as error:
            last_error = error
            time.sleep(pause_seconds)

    raise RuntimeError(
        "All CSV download attempts failed. Last error: "
        + str(last_error)
    )


print("Scraping helper functions are ready.")
````

````python
 # ============================================================
# 4. Download IMF indicators summary catalog
# ============================================================

catalog_download_urls = [
    (
        "https://climatedata.imf.org/api/download/v1/items/"
        + catalog_item_id
        + "/csv?layers=0"
    ),
    (
        "https://climatedata-imf.opendata.arcgis.com/api/download/v1/items/"
        + catalog_item_id
        + "/csv?layers=0"
    ),
    (
        "https://opendata.arcgis.com/api/v3/datasets/"
        + catalog_item_id
        + "_0/downloads/data?format=csv&spatialRefId=4326&where=1%3D1"
    ),
    (
        "https://www.arcgis.com/sharing/rest/content/items/"
        + catalog_item_id
        + "/data"
    )
]

indicators_df, catalog_success_url, catalog_content_bytes = try_read_csv_from_urls(
    catalog_download_urls
)

indicators_df = normalize_columns(indicators_df)

indicators_df.to_csv(
    indicators_summary_file,
    index=False
)

print("Indicators summary downloaded successfully.")
print("Successful URL:")
print(catalog_success_url)

print("Indicators summary shape:")
print(indicators_df.shape)

print("Saved file:")
print(indicators_summary_file)

print("Preview:")
print(indicators_df.head())
````

````python
 # ============================================================
# 5. Check required catalog columns
# ============================================================

required_catalog_columns = [
    "Indicator_ID",
    "Indicator",
    "Country_Coverage",
    "Download_Data_URL"
]

missing_columns = [
    col
    for col in required_catalog_columns
    if col not in indicators_df.columns
]

if len(missing_columns) > 0:
    raise ValueError(
        "The indicators summary is missing required columns: "
        + ", ".join(missing_columns)
    )

print("Required catalog columns are available.")
````

````python
 # ============================================================
# 6. Filter catalog to country-level IMF Climate datasets
# ============================================================

# Keep country-level datasets.
# Remove datasets marked as Regional or Global.
selected_indicators_df = indicators_df[
    ~indicators_df["Country_Coverage"]
    .astype(str)
    .str.strip()
    .str.lower()
    .isin([
        "regional",
        "global"
    ])
].copy()

selected_indicators_df = selected_indicators_df.reset_index(drop=True)

print("Total indicators in catalog:")
print(len(indicators_df))

print("Selected country-level indicators:")
print(len(selected_indicators_df))

print("Preview of selected datasets:")
print(
    selected_indicators_df[
        [
            "Indicator_ID",
            "Indicator",
            "Country_Coverage",
            "Download_Data_URL"
        ]
    ].head()
)
````

````python
 # ============================================================
# 7. Download each selected country-level dataset
# ============================================================

download_manifest_rows = []
downloaded_file_paths = []

for row_index, row in tqdm(
    selected_indicators_df.iterrows(),
    total=len(selected_indicators_df),
    desc="Downloading IMF Climate country-level datasets"
):
    indicator_id = int(row["Indicator_ID"])
    indicator_name = clean_text(row["Indicator"])
    country_coverage = clean_text(row.get("Country_Coverage", ""))
    time_coverage = clean_text(row.get("Time_Coverage", ""))
    download_data_url = clean_text(row.get("Download_Data_URL", ""))

    item_id = extract_item_id_from_url(download_data_url)

    if item_id == "":
        download_manifest_rows.append({
            "Indicator_ID": indicator_id,
            "Indicator": indicator_name,
            "Country_Coverage": country_coverage,
            "Time_Coverage": time_coverage,
            "Download_Data_URL": download_data_url,
            "ArcGIS_Item_ID": "",
            "Downloaded_File": "",
            "Download_Status": "failed_no_item_id",
            "Successful_URL": ""
        })
        continue

    dataset_download_urls = [
        (
            "https://climatedata.imf.org/api/download/v1/items/"
            + item_id
            + "/csv?layers=0"
        ),
        (
            "https://climatedata-imf.opendata.arcgis.com/api/download/v1/items/"
            + item_id
            + "/csv?layers=0"
        ),
        (
            "https://opendata.arcgis.com/api/v3/datasets/"
            + item_id
            + "_0/downloads/data?format=csv&spatialRefId=4326&where=1%3D1"
        ),
        (
            "https://www.arcgis.com/sharing/rest/content/items/"
            + item_id
            + "/data"
        )
    ]

    try:
        dataset_df, success_url, content_bytes = try_read_csv_from_urls(
            dataset_download_urls
        )

        dataset_df = normalize_columns(dataset_df)

        file_name = (
            str(indicator_id).zfill(2)
            + "_"
            + slugify(indicator_name)
            + ".csv"
        )

        dataset_file_path = os.path.join(dataset_folder, file_name)
        raw_file_path = os.path.join(raw_download_folder, file_name)

        # Save cleaned CSV
        dataset_df.to_csv(
            dataset_file_path,
            index=False
        )

        # Save a raw-equivalent CSV copy for backup
        dataset_df.to_csv(
            raw_file_path,
            index=False
        )

        downloaded_file_paths.append(dataset_file_path)

        download_manifest_rows.append({
            "Indicator_ID": indicator_id,
            "Indicator": indicator_name,
            "Country_Coverage": country_coverage,
            "Time_Coverage": time_coverage,
            "Download_Data_URL": download_data_url,
            "ArcGIS_Item_ID": item_id,
            "Downloaded_File": file_name,
            "Download_Status": "success",
            "Successful_URL": success_url
        })

    except Exception as error:
        download_manifest_rows.append({
            "Indicator_ID": indicator_id,
            "Indicator": indicator_name,
            "Country_Coverage": country_coverage,
            "Time_Coverage": time_coverage,
            "Download_Data_URL": download_data_url,
            "ArcGIS_Item_ID": item_id,
            "Downloaded_File": "",
            "Download_Status": "failed_" + str(error)[:250],
            "Successful_URL": ""
        })

    time.sleep(pause_seconds)


manifest_df = pd.DataFrame(download_manifest_rows)

manifest_df.to_csv(
    manifest_file,
    index=False
)

print("Download process completed.")
print("Manifest shape:")
print(manifest_df.shape)

print("Successful downloads:")
print((manifest_df["Download_Status"] == "success").sum())

print("Failed downloads:")
print((manifest_df["Download_Status"] != "success").sum())

print("Manifest file saved:")
print(manifest_file)

print("Manifest preview:")
print(manifest_df.head())
````

````python
 # ============================================================
# 8. Final scraping validation
# ============================================================

downloaded_csv_files = sorted([
    file_name
    for file_name in os.listdir(dataset_folder)
    if file_name.lower().endswith(".csv")
    and file_name != os.path.basename(manifest_file)
])

print("Scraping validation summary")
print("---------------------------")

print("Indicators summary file exists:")
print(os.path.exists(indicators_summary_file))

print("Manifest file exists:")
print(os.path.exists(manifest_file))

print("Number of downloaded dataset CSV files:")
print(len(downloaded_csv_files))

print("First few downloaded files:")
print(downloaded_csv_files[:10])

print("Dataset folder:")
print(dataset_folder)

print("Raw backup folder:")
print(raw_download_folder)

print("Scraping script finished successfully.")
````

````python
 # ============================================================
# 1. Import packages
# ============================================================

import os
import re
import subprocess
import sys
import pandas as pd
import numpy as np
from tqdm.auto import tqdm

# Install country_converter if needed
try:
    import country_converter as coco
except ImportError:
    subprocess.check_call([
        sys.executable,
        "-m",
        "pip",
        "install",
        "country_converter"
    ])
    import country_converter as coco
````

````python
 # ============================================================
# 2. Paths and settings
# ============================================================

# Input paths
template_file = "IMF_final.csv"
data_folder = "imf_climate_country_coverage_datasets"

manifest_file = os.path.join(
    data_folder,
    "imf_climate_country_coverage_manifest.csv"
)

# Output paths
main_output_csv = "IMF_final_variant_country_availability.csv"

detail_output_csv = (
    "IMF_final_variant_country_availability_detail.csv"
)

rules_output_csv = (
    "IMF_final_variant_split_rules.csv"
)

excel_output_file = (
    "IMF_final_variant_country_availability.xlsx"
)

# Year handling
current_year = 2026
latest_year_min_share = 0.25
year_pattern = re.compile(r"^\d{4}$")

print("Data cleaning paths and settings are ready.")
print("Template file:", template_file)
print("Data folder:", data_folder)
print("Manifest file:", manifest_file)
````

````python
# ============================================================
# 3. Helper functions
# ============================================================

def clean_text(value):
    """
    Convert a value to clean text.
    """
    if pd.isna(value):
        return ""

    return re.sub(
        r"\s+",
        " ",
        str(value).strip()
    )


def normalize_columns(df):
    """
    Clean column names and remove unnamed columns.

    This keeps IMF year columns such as F2019, F2020, and F2021 unchanged.
    """
    df = df.copy()

    df.columns = [
        clean_text(col)
        for col in df.columns
    ]

    unnamed_cols = [
        col
        for col in df.columns
        if str(col).lower().startswith("unnamed")
    ]

    if len(unnamed_cols) > 0:
        df = df.drop(columns=unnamed_cols)

    return df


def first_nonempty(series):
    """
    Return the first non-empty value in a pandas Series.
    """
    values = [
        clean_text(value)
        for value in series
        if clean_text(value) != ""
    ]

    if len(values) == 0:
        return ""

    return values[0]


def unique_join(series):
    """
    Join unique non-empty text values using semicolon.
    """
    seen = []

    for value in series:
        text = clean_text(value)

        if text != "" and text not in seen:
            seen.append(text)

    return "; ".join(seen)


def is_year_column(column_name):
    """
    Identify year columns.

    Handles both formats:
      2019
      F2019
    """
    column_text = str(column_name).strip()

    return re.match(r"^F?\d{4}$", column_text) is not None


def extract_year_from_column(column_name):
    """
    Extract the numeric year from a year column.

    Examples:
      2019  -> 2019
      F2019 -> 2019
    """
    column_text = str(column_name).strip()

    match = re.search(r"(\d{4})$", column_text)

    if match is None:
        return None

    return int(match.group(1))


def latest_reporting_years(df, min_share=0.25):
    """
    Select the latest two reporting year columns based on unique country coverage.

    This version handles IMF year columns written as:
      2019
      F2019

    It returns the actual dataframe column names, such as:
      ["F2021", "F2022"]

    This is important because the later code must use the real column names
    when checking whether each country has data.
    """

    year_columns = [
        column
        for column in df.columns
        if is_year_column(column)
    ]

    if len(year_columns) == 0:
        return []

    work_df = df.copy()

    # Create country identifier for coverage counting.
    if "ISO3" in work_df.columns:
        work_df["_country_id_for_year_check"] = (
            work_df["ISO3"]
            .astype(str)
            .str.strip()
            .str.upper()
        )
    elif "Country" in work_df.columns:
        work_df["_country_id_for_year_check"] = (
            work_df["Country"]
            .astype(str)
            .str.strip()
        )
    else:
        return []

    work_df["_country_id_for_year_check"] = (
        work_df["_country_id_for_year_check"]
        .replace(["", "nan", "None", "NAN", "NONE"], np.nan)
    )

    rows = []

    for column in year_columns:
        year_value = extract_year_from_column(column)

        if year_value is None:
            continue

        numeric_values = pd.to_numeric(
            work_df[column],
            errors="coerce"
        )

        available_country_count = (
            work_df.loc[
                numeric_values.notna(),
                "_country_id_for_year_check"
            ]
            .dropna()
            .nunique()
        )

        if available_country_count > 0:
            rows.append({
                "column": column,
                "year": year_value,
                "available_country_count": int(available_country_count)
            })

    counts_df = pd.DataFrame(rows)

    if counts_df.empty:
        return []

    counts_df = counts_df[
        counts_df["year"] <= current_year
    ].copy()

    if counts_df.empty:
        return []

    max_country_count = counts_df["available_country_count"].max()

    eligible_df = counts_df[
        counts_df["available_country_count"] >= max_country_count * min_share
    ].copy()

    if eligible_df.empty:
        eligible_df = counts_df.copy()

    latest_columns = (
        eligible_df
        .sort_values("year")["column"]
        .tail(2)
        .tolist()
    )

    return latest_columns
````

````python
# ============================================================
# 4. Load template, optional manifest, and dataset file list
# ============================================================

# ----------------------------
# 4.1 Check input paths
# ----------------------------

if not os.path.exists(template_file):
    raise FileNotFoundError(
        "Template file not found: "
        + template_file
        + "\nCurrent working directory: "
        + os.getcwd()
    )

if not os.path.exists(data_folder):
    raise FileNotFoundError(
        "Data folder not found: "
        + data_folder
        + "\nRun the scraping script first."
        + "\nCurrent working directory: "
        + os.getcwd()
    )


# ----------------------------
# 4.2 Load and clean template
# ----------------------------

template_df = pd.read_csv(
    template_file,
    low_memory=False
)

template_df = normalize_columns(template_df)

# Standardize the first five expected columns if needed.
# This avoids errors caused by small naming differences.
required_template_columns = [
    "no",
    "indicator",
    "Definition",
    "unit",
    "source"
]

if len(template_df.columns) <= 5:
    raise ValueError(
        "Template file must have at least 6 columns: "
        "no, indicator, Definition, unit, source, and country columns."
    )

# Rename the first five columns to the expected names.
# This assumes your template structure is correct, even if names vary slightly.
template_columns = list(template_df.columns)
template_columns[:5] = required_template_columns
template_df.columns = template_columns

country_columns = list(template_df.columns[5:])

if len(country_columns) == 0:
    raise ValueError(
        "No country columns found in the template. "
        "The template must have country columns after the first five columns."
    )


# ----------------------------
# 4.3 Load manifest if available
# ----------------------------

if os.path.exists(manifest_file):
    manifest_df = pd.read_csv(
        manifest_file,
        low_memory=False
    )

    manifest_df = normalize_columns(manifest_df)

    print("Manifest file loaded:")
    print(manifest_file)

    print("Manifest shape:")
    print(manifest_df.shape)

else:
    manifest_df = pd.DataFrame()

    print("Warning: Manifest file was not found.")
    print("The code will continue using all dataset CSV files in the data folder.")
    print("Expected manifest path:")
    print(manifest_file)


# ----------------------------
# 4.4 List downloaded dataset CSV files
# ----------------------------

excluded_csv_names = {
    os.path.basename(manifest_file),
    os.path.basename(main_output_csv),
    os.path.basename(detail_output_csv),
    os.path.basename(rules_output_csv),
    os.path.basename(excel_output_file).replace(".xlsx", ".csv"),
    "00_Indicators_Summary.csv"
}

csv_files = sorted([
    file_name
    for file_name in os.listdir(data_folder)
    if file_name.lower().endswith(".csv")
    and file_name not in excluded_csv_names
])

if len(csv_files) == 0:
    raise FileNotFoundError(
        "No dataset CSV files found in: "
        + data_folder
        + "\nRun the scraping script first or check the data folder path."
    )


# ----------------------------
# 4.5 Optional: keep only successful downloads from manifest
# ----------------------------

if not manifest_df.empty:
    if (
        "Downloaded_File" in manifest_df.columns
        and "Download_Status" in manifest_df.columns
    ):
        successful_files = (
            manifest_df.loc[
                manifest_df["Download_Status"].astype(str).str.lower().eq("success"),
                "Downloaded_File"
            ]
            .dropna()
            .astype(str)
            .tolist()
        )

        successful_files = [
            file_name
            for file_name in successful_files
            if file_name in csv_files
        ]

        if len(successful_files) > 0:
            csv_files = sorted(successful_files)

            print("Using successful downloads listed in manifest.")
            print("Successful dataset files used:")
            print(len(csv_files))

        else:
            print("Warning: Manifest found, but no successful downloaded files matched the folder.")
            print("The code will continue using all dataset CSV files found in the folder.")


# ----------------------------
# 4.6 Final checks and preview
# ----------------------------

print("Template shape:")
print(template_df.shape)

print("Template first five columns:")
print(list(template_df.columns[:5]))

print("Number of country columns:")
print(len(country_columns))

print("First few country columns:")
print(country_columns[:10])

print("Number of dataset files:")
print(len(csv_files))

print("First few dataset files:")
print(csv_files[:10])
````

````python
# ============================================================
# 5. Define fields excluded from variant split rules
# ============================================================

# These columns identify reporting countries or counterpart countries.
# They are used for country matching and availability assessment,
# but they should not create separate indicator variants.
country_like_columns = {
    "Country",
    "ISO2",
    "ISO3",
    "Country_Name",
    "Country Code",
    "Country_Code",
    "Counterpart Country",
    "Counterpart_Country",
    "Counterpart ISO2",
    "Counterpart_ISO2",
    "Counterpart ISO3",
    "Counterpart_ISO3"
}

# These are metadata fields.
# They describe the dataset, source, unit, or database structure.
# They should not create separate indicator variants.
metadata_columns = country_like_columns | {
    "ObjectId",
    "OBJECTID",
    "FID",
    "Source",
    "Copyright and Usage",
    "Copyright_and_Usage",
    "CTS Code",
    "CTS_Code",
    "CTS Name",
    "CTS_Name",
    "CTS Full Descriptor",
    "CTS_Full_Descriptor",
    "Unit",
    "Scale"
}

print("Columns excluded from variant split rules:")
print(sorted(metadata_columns))
````

````python
# ============================================================
# 6. Country name to ISO3 mapping
# ============================================================

# Manual fixes for country names that often do not map cleanly.
# These are used mainly for template country names.
manual_iso_map = {
    "Congo Dem. Rep.": "COD",
    "Congo, Dem. Rep.": "COD",
    "Democratic Republic of the Congo": "COD",
    "Congo Rep.": "COG",
    "Congo, Rep.": "COG",
    "Republic of Congo": "COG",
    "Côte dIvoire": "CIV",
    "Côte d'Ivoire": "CIV",
    "Cote dIvoire": "CIV",
    "Cote d'Ivoire": "CIV",
    "Hong Kong (S.A.R.)": "HKG",
    "Hong Kong SAR": "HKG",
    "Hong Kong SAR, China": "HKG",
    "Korea Dem. Peoples Rep.": "PRK",
    "Korea, Dem. People's Rep.": "PRK",
    "North Korea": "PRK",
    "South Korea": "KOR",
    "Korea, Rep.": "KOR",
    "Kyrgyz Republic": "KGZ",
    "Lao PDR": "LAO",
    "Lao People's Democratic Republic": "LAO",
    "Macao SAR": "MAC",
    "Macao SAR, China": "MAC",
    "Micronesia": "FSM",
    "Micronesia, Fed. Sts.": "FSM",
    "Russia": "RUS",
    "Russian Federation": "RUS",
    "São Tomé and Príncipe": "STP",
    "Sao Tome and Principe": "STP",
    "Slovak Republic": "SVK",
    "Slovakia": "SVK",
    "St. Kitts and Nevis": "KNA",
    "Saint Kitts and Nevis": "KNA",
    "St. Lucia": "LCA",
    "Saint Lucia": "LCA",
    "St. Vincent and the Grenadines": "VCT",
    "Saint Vincent and the Grenadines": "VCT",
    "Syria": "SYR",
    "Syrian Arab Republic": "SYR",
    "Taiwan Province of China": "TWN",
    "Taiwan": "TWN",
    "Türkiye": "TUR",
    "Turkey": "TUR",
    "United States": "USA",
    "United States of America": "USA",
    "Venezuela": "VEN",
    "Venezuela, RB": "VEN",
    "Vietnam": "VNM",
    "Viet Nam": "VNM",
    "Yemen": "YEM",
    "Yemen, Rep.": "YEM",
    "West Bank and Gaza": "PSE",
    "Palestine": "PSE",
    "Czech Republic": "CZE",
    "Czechia": "CZE",
    "Eswatini": "SWZ",
    "Swaziland": "SWZ",
    "Iran": "IRN",
    "Iran, Islamic Rep.": "IRN",
    "Tanzania": "TZA",
    "Tanzania, United Republic of": "TZA",
    "The Bahamas": "BHS",
    "Bahamas": "BHS",
    "Bahamas, The": "BHS",
    "The Gambia": "GMB",
    "Gambia": "GMB",
    "Gambia, The": "GMB",
    "Brunei Darussalam": "BRN",
    "Brunei": "BRN",
    "Cabo Verde": "CPV",
    "Cape Verde": "CPV",
    "Timor-Leste": "TLS",
    "Timor Leste": "TLS",
    "Channel Islands": "CHI"
}


def clean_iso3(value):
    """
    Clean ISO3 values.

    Returns None for missing or invalid-looking values.
    """
    iso_text = clean_text(value).upper()

    if iso_text in ["", "NAN", "NONE", "NULL", "NOT FOUND"]:
        return None

    if len(iso_text) != 3:
        return None

    if not iso_text.isalpha():
        return None

    return iso_text


def country_to_iso3(country_name):
    """
    Convert country name to ISO3 code.

    This first checks manual fixes, then uses country_converter.
    """
    country_clean = clean_text(country_name)

    if country_clean == "":
        return None

    # If the value is already an ISO3 code, keep it.
    possible_iso3 = clean_iso3(country_clean)
    if possible_iso3 is not None:
        return possible_iso3

    if country_clean in manual_iso_map:
        return manual_iso_map[country_clean]

    iso_value = coco.convert(
        names=country_clean,
        to="ISO3",
        not_found=None
    )

    if isinstance(iso_value, list):
        if len(iso_value) == 0:
            return None
        iso_value = iso_value[0]

    if iso_value is None:
        return None

    iso_text = clean_iso3(iso_value)

    return iso_text


template_country_map_df = pd.DataFrame({
    "country": country_columns
})

template_country_map_df["iso3"] = (
    template_country_map_df["country"]
    .map(country_to_iso3)
)

template_iso_to_country = (
    template_country_map_df
    .dropna(subset=["iso3"])
    .drop_duplicates("iso3")
    .set_index("iso3")["country"]
    .to_dict()
)

unmatched_template_countries = (
    template_country_map_df[
        template_country_map_df["iso3"].isna()
    ]["country"]
    .tolist()
)

print("Template countries mapped to ISO3:")
print(len(template_iso_to_country))

print("Total template country columns:")
print(len(country_columns))

if len(unmatched_template_countries) > 0:
    print("Warning: Some template countries were not mapped to ISO3:")
    print(unmatched_template_countries[:30])

# Save country mapping diagnostic for review.
template_country_map_output = "IMF_template_country_iso3_mapping.csv"

template_country_map_df.to_csv(
    template_country_map_output,
    index=False
)

print("Template country mapping diagnostic saved:")
print(template_country_map_output)
````

````python
# ============================================================
# 7. Inspect split rules before processing
# ============================================================

split_rule_preview_rows = []

for file_name in tqdm(csv_files, desc="Inspecting split rules"):
    file_path = os.path.join(data_folder, file_name)

    df = pd.read_csv(file_path, low_memory=False)
    df = normalize_columns(df)

    # --------------------------------------------------------
    # 7.1 Identify year and non-year columns
    # --------------------------------------------------------

    year_columns = [
        column
        for column in df.columns
        if is_year_column(column)
    ]

    non_year_columns = [
        column
        for column in df.columns
        if column not in year_columns
    ]

    # --------------------------------------------------------
    # 7.2 Identify candidate variant split columns
    # --------------------------------------------------------

    variant_columns = [
        column
        for column in non_year_columns
        if column not in metadata_columns
    ]

    # Keep Indicator first when present.
    if "Indicator" in df.columns:
        variant_columns = (
            ["Indicator"]
            + [
                column
                for column in variant_columns
                if column != "Indicator"
            ]
        )

    # Drop fully empty variant columns.
    clean_variant_columns = []

    for column in variant_columns:
        if column in df.columns:
            nonempty_count = (
                df[column]
                .map(clean_text)
                .replace("", np.nan)
                .notna()
                .sum()
            )

            if nonempty_count > 0:
                clean_variant_columns.append(column)

    variant_columns = clean_variant_columns

    # --------------------------------------------------------
    # 7.3 Identify unit, description, and source fields
    # --------------------------------------------------------

    if "Scale" in df.columns:
        unit_source = "Scale"
    elif "Unit" in df.columns:
        unit_source = "Unit"
    else:
        unit_source = None

    if "CTS Full Descriptor" in df.columns:
        description_source = "CTS Full Descriptor"
    elif "CTS_Full_Descriptor" in df.columns:
        description_source = "CTS_Full_Descriptor"
    elif "Indicator" in df.columns:
        description_source = "Indicator"
    else:
        description_source = None

    if "Source" in df.columns:
        source_source = "Source"
    else:
        source_source = None

    # --------------------------------------------------------
    # 7.4 Inspect latest reporting columns
    # --------------------------------------------------------

    latest_columns = latest_reporting_years(
        df,
        min_share=latest_year_min_share
    )

    # --------------------------------------------------------
    # 7.5 Add preview row
    # --------------------------------------------------------

    split_rule_preview_rows.append({
        "file_name": file_name,
        "variant_columns_used": " + ".join(variant_columns),
        "unit_source": unit_source,
        "description_source": description_source,
        "source_source": source_source,
        "year_columns_found": len(year_columns),
        "first_year_column": year_columns[0] if len(year_columns) > 0 else "",
        "last_year_column": year_columns[-1] if len(year_columns) > 0 else "",
        "latest_reporting_years_used": ", ".join(latest_columns),
        "number_of_variant_columns": len(variant_columns)
    })

split_rule_preview_df = pd.DataFrame(split_rule_preview_rows)

print("Split rule preview:")
print(split_rule_preview_df.to_string(index=False))
````

````python
# ============================================================
# 8. Build variant-level country availability table
# ============================================================

variant_rows = []
variant_detail_rows = []
variant_rules_rows = []
skipped_dataset_rows = []

for file_name in tqdm(csv_files, desc="Building variant availability"):
    file_path = os.path.join(data_folder, file_name)

    df = pd.read_csv(file_path, low_memory=False)
    df = normalize_columns(df)

    # --------------------------------------------------------
    # 8.1 Identify year and non-year columns
    # --------------------------------------------------------

    year_columns = [
        column
        for column in df.columns
        if is_year_column(column)
    ]

    non_year_columns = [
        column
        for column in df.columns
        if column not in year_columns
    ]

    # --------------------------------------------------------
    # 8.2 Identify variant split columns
    # --------------------------------------------------------

    variant_columns = [
        column
        for column in non_year_columns
        if column not in metadata_columns
    ]

    # Keep Indicator first when present.
    if "Indicator" in df.columns:
        variant_columns = (
            ["Indicator"]
            + [
                column
                for column in variant_columns
                if column != "Indicator"
            ]
        )

    # Drop fully empty variant columns.
    clean_variant_columns = []

    for column in variant_columns:
        if column in df.columns:
            nonempty_count = (
                df[column]
                .map(clean_text)
                .replace("", np.nan)
                .notna()
                .sum()
            )

            if nonempty_count > 0:
                clean_variant_columns.append(column)

    variant_columns = clean_variant_columns

    # --------------------------------------------------------
    # 8.3 Identify unit, description, and source columns
    # --------------------------------------------------------

    if "Scale" in df.columns:
        unit_column = "Scale"
    elif "Unit" in df.columns:
        unit_column = "Unit"
    else:
        unit_column = None

    if "CTS Full Descriptor" in df.columns:
        description_column = "CTS Full Descriptor"
    elif "CTS_Full_Descriptor" in df.columns:
        description_column = "CTS_Full_Descriptor"
    elif "Indicator" in df.columns:
        description_column = "Indicator"
    else:
        description_column = None

    if "Source" in df.columns:
        source_column = "Source"
    else:
        source_column = None

    # --------------------------------------------------------
    # 8.4 Select latest reporting columns
    # --------------------------------------------------------

    latest_columns = latest_reporting_years(
        df,
        min_share=latest_year_min_share
    )

    indicator_id_match = re.match(r"^(\d+)_", file_name)

    if indicator_id_match is not None:
        indicator_id = int(indicator_id_match.group(1))
    else:
        indicator_id = len(variant_rules_rows) + 1

    variant_rules_rows.append({
        "file_name": file_name,
        "variant_columns_used": " + ".join(variant_columns),
        "unit_source": unit_column,
        "description_source": description_column,
        "source_source": source_column if source_column is not None else "",
        "latest_reporting_years_used": ", ".join(latest_columns),
        "year_columns_found": len(year_columns)
    })

    if len(latest_columns) == 0:
        skipped_dataset_rows.append({
            "file_name": file_name,
            "reason": "No valid latest reporting year columns found"
        })
        continue

    valid_latest_columns = [
        column
        for column in latest_columns
        if column in df.columns
    ]

    if len(valid_latest_columns) == 0:
        skipped_dataset_rows.append({
            "file_name": file_name,
            "reason": "Latest reporting year columns not found in dataframe"
        })
        continue

    # --------------------------------------------------------
    # 8.5 Define columns to keep
    # --------------------------------------------------------

    keep_columns = []

    if "Country" in df.columns:
        keep_columns.append("Country")

    if "ISO3" in df.columns:
        keep_columns.append("ISO3")

    if len(keep_columns) == 0:
        skipped_dataset_rows.append({
            "file_name": file_name,
            "reason": "No Country or ISO3 column found"
        })
        continue

    for column in variant_columns:
        if column in df.columns and column not in keep_columns:
            keep_columns.append(column)

    for extra_column in [description_column, unit_column, source_column]:
        if extra_column is not None:
            if extra_column in df.columns and extra_column not in keep_columns:
                keep_columns.append(extra_column)

    work_df = df[
        keep_columns + valid_latest_columns
    ].copy()

    # --------------------------------------------------------
    # 8.6 Mark rows with data in at least one selected latest year
    # --------------------------------------------------------

    work_df["has_latest_data"] = (
        work_df[valid_latest_columns]
        .apply(pd.to_numeric, errors="coerce")
        .notna()
        .any(axis=1)
    )

    work_df = work_df.loc[
        work_df["has_latest_data"]
    ].copy()

    if work_df.empty:
        skipped_dataset_rows.append({
            "file_name": file_name,
            "reason": "No rows had data in the selected latest reporting years"
        })
        continue

    # --------------------------------------------------------
    # 8.7 Match reporting country to template country
    # --------------------------------------------------------

    # Country availability uses the reporting country, not counterpart country.
    if "ISO3" in work_df.columns:
        work_df["matched_iso3"] = (
            work_df["ISO3"]
            .map(clean_iso3)
        )

    elif "Country" in work_df.columns:
        work_df["matched_iso3"] = (
            work_df["Country"]
            .map(country_to_iso3)
        )

    else:
        work_df["matched_iso3"] = None

    work_df = work_df.loc[
        work_df["matched_iso3"].isin(template_iso_to_country.keys())
    ].copy()

    if work_df.empty:
        skipped_dataset_rows.append({
            "file_name": file_name,
            "reason": "No reporting countries matched template countries"
        })
        continue

    # --------------------------------------------------------
    # 8.8 Build variant indicator label
    # --------------------------------------------------------

    for column in variant_columns:
        if column in work_df.columns:
            work_df[column] = work_df[column].map(clean_text)

    if len(variant_columns) > 0:
        work_df["variant_indicator"] = (
            work_df[variant_columns]
            .astype(str)
            .agg(" | ".join, axis=1)
        )

        work_df["variant_indicator"] = (
            work_df["variant_indicator"]
            .map(clean_text)
        )

    else:
        work_df["variant_indicator"] = "Dataset " + str(indicator_id)

    # --------------------------------------------------------
    # 8.9 Create one output row per indicator variant
    # --------------------------------------------------------

    variant_number = 0

    for variant_key, group_df in work_df.groupby(
        "variant_indicator",
        dropna=False
    ):
        variant_number = variant_number + 1

        output_row = {
            column: 0
            for column in template_df.columns
        }

        output_row["no"] = (
            str(indicator_id).zfill(2)
            + "-"
            + str(variant_number).zfill(4)
        )

        output_row["indicator"] = variant_key

        if (
            description_column is not None
            and description_column in group_df.columns
        ):
            output_row["Definition"] = first_nonempty(
                group_df[description_column]
            )
        else:
            output_row["Definition"] = variant_key

        if (
            unit_column is not None
            and unit_column in group_df.columns
        ):
            output_row["unit"] = first_nonempty(
                group_df[unit_column]
            )
        else:
            output_row["unit"] = ""

        if (
            source_column is not None
            and source_column in group_df.columns
        ):
            output_row["source"] = unique_join(
                group_df[source_column]
            )
        else:
            output_row["source"] = "IMF Climate Data"

        if output_row["source"] == "":
            output_row["source"] = "IMF Climate Data"

        # Mark countries with available data.
        available_iso3_values = set(
            group_df["matched_iso3"]
            .dropna()
            .astype(str)
        )

        for iso3_value in available_iso3_values:
            country_name = template_iso_to_country.get(iso3_value)

            if country_name in output_row:
                output_row[country_name] = 1

        available_country_count = int(
            sum(
                output_row[country_name]
                for country_name in country_columns
            )
        )

        variant_rows.append(output_row)

        variant_detail_rows.append({
            "no": output_row["no"],
            "dataset_no": indicator_id,
            "file_name": file_name,
            "indicator": output_row["indicator"],
            "Definition": output_row["Definition"],
            "unit": output_row["unit"],
            "source": output_row["source"],
            "variant_columns_used": " + ".join(variant_columns),
            "latest_reporting_years_used": ", ".join(valid_latest_columns),
            "available_country_count": available_country_count
        })
````

````python
# ============================================================
# 9. Create final DataFrames
# ============================================================

final_availability_df = pd.DataFrame(
    variant_rows,
    columns=template_df.columns
)

if not final_availability_df.empty:
    final_availability_df = (
        final_availability_df
        .sort_values("no")
        .reset_index(drop=True)
    )

variant_detail_df = pd.DataFrame(variant_detail_rows)

if not variant_detail_df.empty:
    variant_detail_df = (
        variant_detail_df
        .sort_values("no")
        .reset_index(drop=True)
    )

split_rules_df = pd.DataFrame(variant_rules_rows)

skipped_datasets_df = pd.DataFrame(skipped_dataset_rows)

print("Final availability shape:")
print(final_availability_df.shape)

print("Variant detail shape:")
print(variant_detail_df.shape)

print("Split rules shape:")
print(split_rules_df.shape)

print("Skipped datasets shape:")
print(skipped_datasets_df.shape)

print("Preview:")
print(final_availability_df.head())
````

### 2. COMBINING DATA SETS

````python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm


# ============================================================
# 1. File paths
# ============================================================

file_paths = {
    'imf': 'IMF_final_variant_country_availability.csv',
    'unsd': 'UNSD_Environmental_Indicators_Metadata_Availability.xlsx',
    'wb': 'WB_final_CWON_full_indicator_variations.csv'
}


# ============================================================
# 2. Settings
# ============================================================

meta_cols = [
    'no',
    'indicator',
    'Definition',
    'unit',
    'source'
]

source_prefix = {
    'wb': 'wb',
    'unsd': 'un',
    'imf': 'imf'
}

ordered_sources = [
    'wb',
    'unsd',
    'imf'
]


# ============================================================
# 3. Load datasets
# ============================================================

loaded = {}

for source_name, path_val in tqdm(file_paths.items(), desc='Loading files'):
    if path_val.endswith('.csv'):
        loaded[source_name] = pd.read_csv(
            path_val,
            encoding='utf-8'
        )

    elif path_val.endswith('.xlsx'):
        loaded[source_name] = pd.read_excel(path_val)

    else:
        raise ValueError('Unsupported file type: ' + path_val)


# ============================================================
# 4. Inspect loaded data
# ============================================================

for source_name, df_val in loaded.items():
    print('\nDataset:')
    print(source_name)

    print('Shape:')
    print(df_val.shape)

    print('Preview:')
    display(df_val.head())


# ============================================================
# 5. Identify country columns
# ============================================================

country_cols_by_source = {}

for source_name, df_val in loaded.items():
    country_cols_by_source[source_name] = [
        col_val for col_val in df_val.columns
        if col_val not in meta_cols
    ]

base_country_cols = country_cols_by_source['imf']


# ============================================================
# 6. Check whether country columns match across datasets
# ============================================================

alignment_rows = []

for source_name, country_cols_val in country_cols_by_source.items():
    alignment_rows.append({
        'dataset': source_name,
        'rows': len(loaded[source_name]),
        'total_columns': len(loaded[source_name].columns),
        'country_columns': len(country_cols_val),
        'same_country_order_as_imf': country_cols_val == base_country_cols,
        'missing_vs_imf': len(set(base_country_cols) - set(country_cols_val)),
        'extra_vs_imf': len(set(country_cols_val) - set(base_country_cols))
    })

alignment_df = pd.DataFrame(alignment_rows)

print('\nCountry column alignment check:')
display(alignment_df)


# ------------------------------------------------------------
# 6A. Stop if country columns do not match
# ------------------------------------------------------------

if not alignment_df['same_country_order_as_imf'].all():
    raise ValueError(
        'Country columns do not match exactly across datasets. '
        'Please inspect alignment_df before combining.'
    )

country_cols = base_country_cols


# ============================================================
# 7. Prefix original IDs and combine datasets vertically
# ============================================================

combined_parts = []

for source_name in tqdm(ordered_sources, desc='Combining datasets'):
    df_part = loaded[source_name].copy()

    prefix_val = source_prefix[source_name]

    # Store prefixed source-specific ID in the original column
    df_part['original'] = [
        prefix_val + '_' + str(row_idx + 1).zfill(4)
        for row_idx in range(len(df_part))
    ]

    # Add source dataset label
    df_part['dataset'] = source_name

    # Reorder columns temporarily.
    # The final no column will be created after combining all datasets.
    df_part = df_part[
        [
            'indicator',
            'Definition',
            'unit',
            'source',
            'original',
            'dataset'
        ]
        + country_cols
    ]

    combined_parts.append(df_part)

combined_wide_df = pd.concat(
    combined_parts,
    ignore_index=True
)


# ------------------------------------------------------------
# 7A. Add final sequential no column
# ------------------------------------------------------------

combined_wide_df.insert(
    0,
    'no',
    range(1, len(combined_wide_df) + 1)
)


# ============================================================
# 8. Validate combined wide dataset
# ============================================================

validation_summary = pd.DataFrame({
    'check': [
        'total_rows',
        'total_columns',
        'country_columns',
        'unique_no_values',
        'duplicate_no_values',
        'unique_original_values',
        'duplicate_original_values'
    ],
    'value': [
        len(combined_wide_df),
        len(combined_wide_df.columns),
        len(country_cols),
        combined_wide_df['no'].nunique(),
        combined_wide_df['no'].duplicated().sum(),
        combined_wide_df['original'].nunique(),
        combined_wide_df['original'].duplicated().sum()
    ]
})

source_summary = (
    combined_wide_df
    .groupby('dataset')
    .agg(
        rows=('no', 'count'),
        first_no=('no', 'min'),
        last_no=('no', 'max'),
        unique_original=('original', 'nunique')
    )
    .reset_index()
)

print('\nValidation summary:')
display(validation_summary)

print('\nSource summary:')
display(source_summary)

print('\nCombined wide preview:')
display(combined_wide_df.head())


# ============================================================
# 9. Create availability summary
# ============================================================

availability_summary_rows = []

for dataset_name, df_group in tqdm(
    combined_wide_df.groupby('dataset'),
    desc='Summarizing availability'
):
    country_values = df_group[country_cols]

    available_cells = int((country_values == 1).sum().sum())
    total_country_cells = int(country_values.size)

    availability_summary_rows.append({
        'dataset': dataset_name,
        'indicators': len(df_group),
        'country_columns': len(country_cols),
        'available_cells': available_cells,
        'total_country_cells': total_country_cells,
        'availability_rate': available_cells / total_country_cells
    })

availability_summary_df = pd.DataFrame(availability_summary_rows)

print('\nAvailability summary:')
display(availability_summary_df)


# ============================================================
# 10. Save outputs
# ============================================================

wide_file = 'combined_country_availability_wide.csv'
excel_file = 'combined_country_availability.xlsx'


# ------------------------------------------------------------
# 10A. Save CSV output
# ------------------------------------------------------------

combined_wide_df.to_csv(
    wide_file,
    index=False,
    encoding='utf-8-sig'
)


# ------------------------------------------------------------
# 10B. Save Excel output
# ------------------------------------------------------------

with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
    combined_wide_df.to_excel(
        writer,
        sheet_name='wide_country_columns',
        index=False
    )

    validation_summary.to_excel(
        writer,
        sheet_name='validation_summary',
        index=False
    )

    source_summary.to_excel(
        writer,
        sheet_name='source_summary',
        index=False
    )

    alignment_df.to_excel(
        writer,
        sheet_name='country_alignment',
        index=False
    )

    availability_summary_df.to_excel(
        writer,
        sheet_name='availability_summary',
        index=False
    )


# ------------------------------------------------------------
# 10C. Confirm saved files
# ------------------------------------------------------------

print('\nSaved files:')
print(wide_file)
print(excel_file)
````

### 3. FILTERING
#### a. Method for Identifying Potential Indicator Replication and Cross-Source Overlap

**Purpose**

This workflow identifies potential replication, duplication, and conceptual overlap among indicators in a combined country-availability dataset. The method is designed for **human review**, not automatic deletion. It creates comparison files that help reviewers decide whether two indicators should both be retained, merged conceptually, or one should be excluded.

The procedure supports two review tasks:

1. **Within-source replication**: potential duplicates within the same source family, such as `imf` versus `imf`.
2. **Cross-source overlap**: potential overlap across source families, such as `imf` versus `wb`, `imf` versus `unsd`, or `unsd` versus `wb`.

This document focuses especially on recreating the file:

```text
cross_source_overlap_candidates.csv
```

**Core Logic**

The workflow compares each pair of indicators using four main dimensions:

| Dimension | Measure | Purpose |
|---|---|---|
| Indicator text | `indicator_similarity_no_timing` | Compares indicator names after timing words and years are removed |
| Indicator plus definition | `text_similarity_no_timing` | Compares broader meaning using indicator name plus definition |
| Country coverage | `coverage_jaccard` | Measures overlap in available country sets |
| Country coverage containment | `coverage_containment` | Checks whether the smaller country set is mostly contained in the larger one |

The method also reports:

| Field | Purpose |
|---|---|
| `unit_match` | Checks whether the unit values are identical |
| `shared_terms` | Shows which substantive words are shared |
| `timing_terms_a`, `timing_terms_b` | Shows whether the difference may be mainly timing-related |
| `review_note` | Reminds reviewers that the pair requires human judgment |


*Define terms to remove*

Timing words and years are removed to detect indicators that differ mainly by measurement period, such as annual, latest year, baseline, target, or projection.

timing_terms = {
    "year", "years", "annual", "annually",
    "latest", "recent", "current", "historical",
    "time", "period", "periods",
    "baseline", "base",
    "projected", "projection", "projections",
    "forecast", "forecasts",
    "target", "targets",
    "estimate", "estimates", "estimated",
    "measured", "measurement",
    "data"
}

stop_terms = {
    "the", "and", "or", "of", "in", "to", "for", "from",
    "by", "with", "as", "a", "an", "per",
    "total", "share", "percentage", "percent",
    "number", "amount", "rate", "index",
    "value", "values",
    "country", "countries",
    "indicator", "indicators"
}
 
 *Build TF-IDF matrix for text similarity*

The workflow uses term frequency-inverse document frequency (TF-IDF) on the combined indicator and definition text after timing words are removed. It uses one-word and two-word terms.
*Set cross-source screening thresholds*

These thresholds are intentionally broad. The goal is to produce a human-review file, not an automatic deletion file.

min_indicator_similarity = 0.60
min_text_similarity = 0.55
min_coverage_jaccard = 0.25
min_coverage_containment = 0.70
min_overlap_country_count = 20
min_shared_terms = 2
 
*How to Interpret the Output*

Each row in `cross_source_overlap_candidates.csv` is one possible overlap pair across two source families.

| Column | Interpretation |
|---|---|
| `source_pair` | Source comparison, such as `imf-wb` |
| `indicator_a`, `indicator_b` | Indicator names being compared |
| `definition_a`, `definition_b` | Definitions used for human review |
| `unit_a`, `unit_b` | Units to check whether indicators measure the same quantity |
| `indicator_similarity_no_timing` | Similarity of indicator names after timing terms are removed |
| `text_similarity_no_timing` | TF-IDF cosine similarity of indicator plus definition |
| `coverage_jaccard` | Country overlap divided by country union |
| `coverage_containment` | Country overlap divided by the smaller country set |
| `overlap_country_count` | Number of countries available in both indicators |
| `unit_match` | Whether units match exactly |
| `shared_terms` | Substantive words shared by both indicators |
| `timing_terms_a`, `timing_terms_b` | Timing-related terms found in each indicator |
| `review_note` | Reminder that the pair requires human judgment |

*Important Limitations*

This method does not prove that two indicators are duplicates. It only creates a prioritized screening file.

Final decisions should be based on:

1. Indicator name.
2. Definition.
3. Unit.
4. Source description.
5. Country coverage.
6. Timing of measurement.
7. Whether the indicator measures a stock, flow, action, or policy guide.
8. Whether one indicator is more continuous, more recent, or better documented than the other.



````python
# ============================================================
# Recreate cross-source overlap candidates
# Input:
#   combined_country_availability_wide.csv
#
# Output:
#   cross_source_overlap_candidates_recreated.csv
# ============================================================

import os
import re
import unicodedata
import numpy as np
import pandas as pd
from itertools import combinations
from difflib import SequenceMatcher
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm.auto import tqdm
````

````python
# ============================================================
# 1. Paths
# ============================================================

input_file = "combined_country_availability_wide.csv"

output_file = "cross_source_overlap_candidates_recreated.csv"

print("Input file:", input_file)
print("Output file:", output_file)
````

````python
# ============================================================
# 2. Load data
# ============================================================

df = pd.read_csv(input_file, low_memory=False)

print("Input shape:")
print(df.shape)

print("Columns:")
print(df.columns[:15].tolist())
````

````python
# ============================================================
# 3. Identify metadata and country columns
# ============================================================

metadata_columns = [
    "no",
    "indicator",
    "Definition",
    "unit",
    "source",
    "original",
    "dataset"
]

country_columns = [
    col
    for col in df.columns
    if col not in metadata_columns
]

print("Number of country columns:")
print(len(country_columns))

print("First few country columns:")
print(country_columns[:10])
````

````python
# ============================================================
# 4. Text cleaning helpers
# ============================================================

def clean_text(value):
    """
    Convert text to a clean lowercase string.
    """
    if pd.isna(value):
        return ""

    return re.sub(
        r"\s+",
        " ",
        str(value).strip()
    )


def normalize_ascii(value):
    """
    Normalize accented characters and remove non-ASCII artifacts.
    """
    value = clean_text(value)

    value = unicodedata.normalize("NFKD", value)
    value = value.encode("ascii", "ignore").decode("ascii")

    return value


timing_terms = {
    "year", "years", "annual", "annually",
    "latest", "recent", "current", "historical",
    "time", "period", "periods",
    "baseline", "base",
    "projected", "projection", "projections",
    "forecast", "forecasts",
    "target", "targets",
    "estimate", "estimates", "estimated",
    "measured", "measurement",
    "data"
}

# Add year-like terms from 1900 to 2100
for year in range(1900, 2101):
    timing_terms.add(str(year))


stop_terms = {
    "the", "and", "or", "of", "in", "to", "for", "from",
    "by", "with", "as", "a", "an", "per",
    "total", "share", "percentage", "percent",
    "number", "amount", "rate", "index",
    "value", "values",
    "country", "countries",
    "indicator", "indicators"
}


def tokenize_text(value, remove_timing=True):
    """
    Convert text to normalized tokens.
    """
    text = normalize_ascii(value).lower()

    text = re.sub(r"[^a-z0-9]+", " ", text)

    tokens = []

    for token in text.split():
        if len(token) <= 1:
            continue

        if remove_timing:
            if token in timing_terms:
                continue

            if re.fullmatch(r"\d{4}", token):
                continue

        if token in stop_terms:
            continue

        tokens.append(token)

    return tokens


def normalized_text(value, remove_timing=True):
    """
    Return normalized text after token cleaning.
    """
    return " ".join(tokenize_text(value, remove_timing=remove_timing))


def token_overlap_similarity(text_a, text_b):
    """
    Similarity based on token overlap.

    Formula:
      overlap / min(number of unique tokens in A, number of unique tokens in B)

    This is useful when one indicator is a shorter version of another.
    Example:
      'land area covered by agricultural area'
      versus
      'land area'
    """
    tokens_a = set(tokenize_text(text_a, remove_timing=True))
    tokens_b = set(tokenize_text(text_b, remove_timing=True))

    if len(tokens_a) == 0 or len(tokens_b) == 0:
        return 0.0

    overlap = len(tokens_a.intersection(tokens_b))
    denominator = min(len(tokens_a), len(tokens_b))

    return overlap / denominator


def shared_terms(text_a, text_b):
    """
    Return shared cleaned terms between two text fields.
    """
    tokens_a = set(tokenize_text(text_a, remove_timing=True))
    tokens_b = set(tokenize_text(text_b, remove_timing=True))

    shared = sorted(tokens_a.intersection(tokens_b))

    return "; ".join(shared)


def extract_timing_terms(value):
    """
    Extract timing-related terms from text.
    """
    tokens = tokenize_text(value, remove_timing=False)

    found = []

    for token in tokens:
        if token in timing_terms or re.fullmatch(r"\d{4}", token):
            if token not in found:
                found.append(token)

    return "; ".join(found)
````

````python
# ============================================================
# 5. Country coverage helpers
# ============================================================

def available_country_set(row, country_columns):
    """
    Return the set of countries coded as available, meaning value == 1.
    """
    available = set()

    for country in country_columns:
        value = row[country]

        try:
            numeric_value = float(value)
        except Exception:
            numeric_value = 0

        if numeric_value == 1:
            available.add(country)

    return available


def coverage_scores(set_a, set_b):
    """
    Compute country-coverage overlap measures.

    coverage_jaccard:
      intersection / union

    coverage_containment:
      intersection / smaller set

    containment is useful when one source covers fewer countries
    but those countries are almost all included in the other source.
    """
    if len(set_a) == 0 and len(set_b) == 0:
        return 0.0, 0.0, 0

    intersection_count = len(set_a.intersection(set_b))
    union_count = len(set_a.union(set_b))

    if union_count == 0:
        jaccard = 0.0
    else:
        jaccard = intersection_count / union_count

    smaller_count = min(len(set_a), len(set_b))

    if smaller_count == 0:
        containment = 0.0
    else:
        containment = intersection_count / smaller_count

    return jaccard, containment, intersection_count
````

````python
# ============================================================
# 6. Prepare comparison fields
# ============================================================

df = df.copy()

df["indicator_clean"] = df["indicator"].map(clean_text)
df["definition_clean"] = df["Definition"].map(clean_text)
df["unit_clean"] = df["unit"].map(clean_text)
df["source_clean"] = df["source"].map(clean_text)

df["indicator_no_timing"] = df["indicator_clean"].map(
    lambda x: normalized_text(x, remove_timing=True)
)

df["full_text_no_timing"] = (
    df["indicator_clean"].fillna("")
    + " "
    + df["definition_clean"].fillna("")
).map(
    lambda x: normalized_text(x, remove_timing=True)
)

df["timing_terms"] = (
    df["indicator_clean"].fillna("")
    + " "
    + df["definition_clean"].fillna("")
).map(extract_timing_terms)

df["available_countries"] = df.apply(
    lambda row: available_country_set(row, country_columns),
    axis=1
)

df["country_count"] = df["available_countries"].map(len)

print("Prepared dataframe shape:")
print(df.shape)

print("Dataset counts:")
print(df["dataset"].value_counts())
````

````python
# ============================================================
# 7. Build TF-IDF similarity matrix for full text
# ============================================================

# TF-IDF is used for indicator + definition similarity.
# The token overlap measure is used separately for indicator-name similarity.

vectorizer = TfidfVectorizer(
    min_df=1,
    ngram_range=(1, 2)
)

tfidf_matrix = vectorizer.fit_transform(
    df["full_text_no_timing"].fillna("")
)

print("TF-IDF matrix shape:")
print(tfidf_matrix.shape)
````

````python
# ============================================================
# 8. Cross-source comparison settings
# ============================================================

# These thresholds are intentionally broad.
# The goal is to generate a human-review file, not automatic deletion.

min_indicator_similarity = 0.60
min_text_similarity = 0.55
min_coverage_jaccard = 0.25
min_coverage_containment = 0.70
min_overlap_country_count = 20
min_shared_terms = 2

print("Cross-source screening thresholds:")
print("min_indicator_similarity:", min_indicator_similarity)
print("min_text_similarity:", min_text_similarity)
print("min_coverage_jaccard:", min_coverage_jaccard)
print("min_coverage_containment:", min_coverage_containment)
print("min_overlap_country_count:", min_overlap_country_count)
print("min_shared_terms:", min_shared_terms)
````

````python
# ============================================================
# 9. Create cross-source overlap candidates
# ============================================================

cross_source_rows = []
pair_id = 0

datasets = sorted(df["dataset"].dropna().unique().tolist())

print("Datasets found:")
print(datasets)

for dataset_a, dataset_b in combinations(datasets, 2):

    df_a = df[df["dataset"] == dataset_a].copy()
    df_b = df[df["dataset"] == dataset_b].copy()

    print("Comparing:", dataset_a, "vs", dataset_b)
    print("Rows:", len(df_a), "x", len(df_b))

    for idx_a, row_a in tqdm(
        df_a.iterrows(),
        total=len(df_a),
        desc=f"{dataset_a} vs {dataset_b}"
    ):
        for idx_b, row_b in df_b.iterrows():

            indicator_similarity = token_overlap_similarity(
                row_a["indicator_clean"],
                row_b["indicator_clean"]
            )

            text_similarity = cosine_similarity(
                tfidf_matrix[idx_a],
                tfidf_matrix[idx_b]
            )[0, 0]

            coverage_jaccard, coverage_containment, overlap_country_count = (
                coverage_scores(
                    row_a["available_countries"],
                    row_b["available_countries"]
                )
            )

            shared = shared_terms(
                row_a["indicator_clean"] + " " + row_a["definition_clean"],
                row_b["indicator_clean"] + " " + row_b["definition_clean"]
            )

            shared_count = len([
                term
                for term in shared.split(";")
                if term.strip() != ""
            ])

            unit_match = (
                "Yes"
                if row_a["unit_clean"].lower() == row_b["unit_clean"].lower()
                and row_a["unit_clean"] != ""
                else "No"
            )

            # Broad cross-source flagging rule.
            # A pair is kept if it has meaningful textual similarity,
            # meaningful country coverage overlap, or strong containment.
            keep_pair = (
                (
                    indicator_similarity >= min_indicator_similarity
                    or text_similarity >= min_text_similarity
                )
                and shared_count >= min_shared_terms
                and (
                    coverage_jaccard >= min_coverage_jaccard
                    or coverage_containment >= min_coverage_containment
                    or overlap_country_count >= min_overlap_country_count
                )
            )

            if not keep_pair:
                continue

            pair_id += 1

            cross_source_rows.append({
                "pair_id": pair_id,
                "source_pair": dataset_a + "-" + dataset_b,

                "dataset_a": row_a["dataset"],
                "original_a": row_a["original"],
                "no_a": row_a["no"],
                "indicator_a": row_a["indicator"],
                "definition_a": row_a["Definition"],
                "unit_a": row_a["unit"],
                "source_a": row_a["source"],
                "country_count_a": row_a["country_count"],

                "dataset_b": row_b["dataset"],
                "original_b": row_b["original"],
                "no_b": row_b["no"],
                "indicator_b": row_b["indicator"],
                "definition_b": row_b["Definition"],
                "unit_b": row_b["unit"],
                "source_b": row_b["source"],
                "country_count_b": row_b["country_count"],

                "indicator_similarity_no_timing": round(indicator_similarity, 3),
                "text_similarity_no_timing": round(float(text_similarity), 3),
                "coverage_jaccard": round(coverage_jaccard, 3),
                "coverage_containment": round(coverage_containment, 3),
                "overlap_country_count": overlap_country_count,
                "unit_match": unit_match,
                "shared_terms": shared,
                "timing_terms_a": row_a["timing_terms"],
                "timing_terms_b": row_b["timing_terms"],
                "review_note": (
                    "Cross-source candidate. Use text similarity plus "
                    "coverage overlap and unit match to decide."
                )
            })

cross_source_df = pd.DataFrame(cross_source_rows)

print("Cross-source candidate shape:")
print(cross_source_df.shape)
````

````python
# ============================================================
# 10. Sort and save cross-source candidates
# ============================================================

if not cross_source_df.empty:
    cross_source_df = (
        cross_source_df
        .sort_values(
            by=[
                "coverage_containment",
                "coverage_jaccard",
                "text_similarity_no_timing",
                "indicator_similarity_no_timing",
                "overlap_country_count"
            ],
            ascending=[
                False,
                False,
                False,
                False,
                False
            ]
        )
        .reset_index(drop=True)
    )

    cross_source_df["pair_id"] = range(1, len(cross_source_df) + 1)

cross_source_df.to_csv(
    output_file,
    index=False
)

print("Saved:")
print(output_file)

print("Final shape:")
print(cross_source_df.shape)

print("Preview:")
print(cross_source_df.head())
````

#### b.  Human Review and Exclusion Decision

After generating the automatic similarity assessment, we used the results as a **screening tool for human review**, not as an automatic deletion rule. The automatic assessment provided similarity scores, country-coverage overlap, unit comparison, shared terms, and explanatory notes. These outputs helped identify which indicator pairs required closer inspection.

The human review used the file `cross_source_overlap_candidates_evaluated` as the main review document. In this file, each potential overlap pair was assessed using the automatic explanation as supporting evidence. We then checked whether the indicators were truly redundant in meaning, unit, definition, and country coverage. The review was intentionally conservative: indicators were excluded only when they showed almost exact similarity and when retaining both would likely create duplication in the final dataset.

The review focused especially on cases where the apparent difference was limited to measurement timing, wording, or source-specific naming, rather than a substantive difference in the concept being measured.

````python
# ============================================================
# Exclude selected indicators from combined country availability
# ============================================================

import pandas as pd

# Input and output files
input_file = "combined_country_availability_wide.csv"
output_file = "combined_country_availability_final.csv"

# Indicators to exclude
exclude_ids = [
    "un_0038",
    "un_0037",
    "un_0038",
    "un_0037",
    "un_0040"
]

# Keep unique exclusion IDs
exclude_ids = sorted(set(exclude_ids))

# Load data
df = pd.read_csv(input_file, low_memory=False)

print("Original row count:")
print(len(df))

# Check which column contains the indicator IDs
possible_id_columns = ["original", "no", "indicator_id", "id"]

id_column = None

for col in possible_id_columns:
    if col in df.columns:
        matched_count = df[col].astype(str).isin(exclude_ids).sum()

        if matched_count > 0:
            id_column = col
            break

if id_column is None:
    raise ValueError(
        "Could not find the exclusion IDs in columns: "
        + ", ".join(possible_id_columns)
        + "\nAvailable columns are:\n"
        + ", ".join(df.columns)
    )

print("Using ID column:")
print(id_column)

print("Rows to exclude:")
print(df[id_column].astype(str).isin(exclude_ids).sum())

# Exclude rows
df_clean = df.loc[
    ~df[id_column].astype(str).isin(exclude_ids)
].copy()

print("Final row count:")
print(len(df_clean))

print("Rows removed:")
print(len(df) - len(df_clean))

# Save output
df_clean.to_csv(output_file, index=False)

print("Saved cleaned file:")
print(output_file)
````
