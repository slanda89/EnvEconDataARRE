# Create Panel A

```python
from __future__ import annotations
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd


# ============================================================
# Step 1. File names
# ============================================================
 

REGION_INPUT = Path("step2_region_category_average_shares.csv")
INCOME_INPUT = Path("step2_income_category_average_shares.csv")

PANEL_A_PNG = Path("step3_panel_A_region_coverage.png")
PANEL_AX_PNG = Path("step3_panel_Ax_income_coverage.png")
COMBINED_PNG = Path("step3_panel_A_and_Ax_coverage.png")
COMBINED_SVG = Path("step3_panel_A_and_Ax_coverage.svg")
COMBINED_PDF = Path("step3_panel_A_and_Ax_coverage.pdf")


# ============================================================
# Step 2. Category columns and labels
# ============================================================

CATEGORY_COLUMNS = [
    "Physical stock",
    "Monetary stock",
    "Physical flow",
    "Monetary flow",
    "Action_kept",
    "Policy Guide_kept",
    "HPS_monetary_only",
    "DDP_monetary_only",
    "MS-NMS_monetary_only",
    "ADE_monetary_only",
]

CATEGORY_LABELS = {
    "Physical stock": "Physical\nstock",
    "Monetary stock": "Monetary\nstock",
    "Physical flow": "Physical\nflow",
    "Monetary flow": "Monetary\nflow",
    "Action_kept": "Action",
    "Policy Guide_kept": "Policy\nGuide",
    "HPS_monetary_only": "Household-produced\nservices (HPS),\nmonetary only",
    "DDP_monetary_only": "Depreciation or\ndegradation process\n(DDP), monetary only",
    "MS-NMS_monetary_only": "Market services and\nnear-market services\n(MS-NMS), monetary only",
    "ADE_monetary_only": "Avoided defensive\nexpenditure (ADE),\nmonetary only",
}


# ============================================================
# Step 3. Group order and display labels
# ============================================================

REGION_ORDER = [
    "Eurostat + UK",
    "Non-Eurostat Europe & Central Asia",
    "Middle East, North Africa, Afghanistan & Pakistan",
    "Sub-Saharan Africa",
    "South Asia",
    "East Asia & Pacific",
    "North America",
    "Latin America & Caribbean",
]

INCOME_ORDER = [
    "Low income",
    "Lower middle income",
    "Upper middle income",
    "High income",
    "Unclassified income",
]

REGION_LABELS = {
    "Middle East, North Africa, Afghanistan & Pakistan": "MENA,\nAfghanistan & Pakistan",
    "Latin America & Caribbean": "Latin America &\nCaribbean",
    "Europe & Central Asia": "Europe &\nCentral Asia",
    "East Asia & Pacific": "East Asia &\nPacific",
    "Sub-Saharan Africa": "Sub-Saharan\nAfrica",
}


# ============================================================
# Step 4. Load and validate group tables
# ============================================================

def load_group_table(path: Path, group_column: str, order: list[str]) -> pd.DataFrame:
    df = pd.read_csv(path)

    missing = [
        col for col in [group_column, "country_count"] + CATEGORY_COLUMNS
        if col not in df.columns
    ]

    if missing:
        raise ValueError(f"{path.name} is missing required columns: {missing}")

    df[group_column] = df[group_column].fillna("Unclassified income")
    df[group_column] = df[group_column].replace({"nan": "Unclassified income"})

    ordered = (
        df.set_index(group_column)
        .reindex(order)
        .dropna(how="all", subset=CATEGORY_COLUMNS)
        .reset_index()
    )

    return ordered


# ============================================================
# Step 5. Figure helper functions
# ============================================================

def annotate_heatmap(ax, values: pd.DataFrame) -> None:
    for y in range(values.shape[0]):
        for x in range(values.shape[1]):
            value = values.iat[y, x]
            text_color = "white" if value >= 55 else "#1f2933"

            ax.text(
                x,
                y,
                f"{value:.1f}",
                ha="center",
                va="center",
                fontsize=8.5,
                color=text_color,
            )


def draw_panel(
    ax,
    df: pd.DataFrame,
    group_column: str,
    title: str,
    y_label_map: dict[str, str] | None = None,
) -> None:
    values = df[CATEGORY_COLUMNS].astype(float)

    image = ax.imshow(
        values,
        vmin=0,
        vmax=100,
        cmap="Blues",
        aspect="auto",
    )

    x_labels = [CATEGORY_LABELS[col] for col in CATEGORY_COLUMNS]
    y_labels = [str(value) for value in df[group_column]]

    if y_label_map:
        y_labels = [y_label_map.get(value, value) for value in y_labels]

    country_counts = df["country_count"].astype(int).tolist()
    y_labels = [
        f"{label}\n(n={count})"
        for label, count in zip(y_labels, country_counts)
    ]

    ax.set_title(title, loc="left", fontsize=14, fontweight="bold", pad=12)
    ax.set_xticks(range(len(CATEGORY_COLUMNS)))
    ax.set_xticklabels(
        x_labels,
        rotation=35,
        ha="right",
        rotation_mode="anchor",
        fontsize=9,
    )
    ax.set_yticks(range(len(y_labels)))
    ax.set_yticklabels(y_labels, fontsize=9)
    ax.set_xlabel("Indicator category", fontsize=10, labelpad=8)
    ax.set_ylabel(group_column, fontsize=10, labelpad=8)

    ax.set_xticks([x - 0.5 for x in range(1, len(CATEGORY_COLUMNS))], minor=True)
    ax.set_yticks([y - 0.5 for y in range(1, len(y_labels))], minor=True)
    ax.grid(which="minor", color="white", linewidth=1.5)
    ax.tick_params(which="minor", bottom=False, left=False)
    ax.tick_params(axis="both", length=0)

    for spine in ax.spines.values():
        spine.set_visible(False)

    annotate_heatmap(ax, values)

    return image


def save_single_panel(
    df: pd.DataFrame,
    group_column: str,
    title: str,
    output_path: Path,
    y_label_map: dict[str, str] | None = None,
) -> None:
    height = max(4.5, 1.0 + len(df) * 0.75)

    fig, ax = plt.subplots(figsize=(12.5, height), constrained_layout=True)

    image = draw_panel(
        ax,
        df,
        group_column,
        title,
        y_label_map,
    )

    cbar = fig.colorbar(image, ax=ax, fraction=0.025, pad=0.02)
    cbar.set_label("Average country coverage (%)", fontsize=9)

    fig.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.close(fig)


# ============================================================
# Step 6. Run figure creation
# ============================================================

region_df = load_group_table(REGION_INPUT, "Region", REGION_ORDER)
income_df = load_group_table(INCOME_INPUT, "Income", INCOME_ORDER)

save_single_panel(
    region_df,
    "Region",
    "Panel A. Regional country average coverage by indicator category",
    PANEL_A_PNG,
    REGION_LABELS,
)

save_single_panel(
    income_df,
    "Income",
    "Panel Ax. Income-group country average coverage by indicator category",
    PANEL_AX_PNG,
)

fig, axes = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(13.5, 11.5),
    constrained_layout=True,
    gridspec_kw={"height_ratios": [1.25, 1.0]},
)

image = draw_panel(
    axes[0],
    region_df,
    "Region",
    "A. Regional country average coverage by indicator category",
    REGION_LABELS,
)

draw_panel(
    axes[1],
    income_df,
    "Income",
    "Ax. Income-group country average coverage by indicator category",
)

cbar = fig.colorbar(image, ax=axes, fraction=0.018, pad=0.015)
cbar.set_label("Average country coverage (%)", fontsize=10)

fig.savefig(COMBINED_PNG, dpi=300, bbox_inches="tight")
fig.savefig(COMBINED_SVG, bbox_inches="tight")
fig.savefig(COMBINED_PDF, bbox_inches="tight")

plt.close(fig)

print("Step 3 Panel A and Ax figures complete")
print(f"Panel A: {PANEL_A_PNG}")
print(f"Panel Ax: {PANEL_AX_PNG}")
print(f"Combined PNG: {COMBINED_PNG}")
print(f"Combined SVG: {COMBINED_SVG}")
print(f"Combined PDF: {COMBINED_PDF}")
```
